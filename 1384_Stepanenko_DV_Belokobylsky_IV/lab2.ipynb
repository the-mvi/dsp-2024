## Основные теоретические положения

$s_1(k) = \begin{cases} U,& \quad n_0 \leqslant k \leqslant n_0 + n_{imp} - 1; \\ 0,& \quad \mbox{иначе} \\ \end{cases}$ - дискретный прямоугольный импульс

$s_2(k) = a_1x_1(k) + a_2x_2(k) + a_3x_3(k)$ - линейная комбинация дискретных гармонических сигналов, где $x_i(k) = B_i\sin(\hat\omega_ik), \quad i = 1, 2, 3$


$\overline{s(k)}=\frac{\sum_{k=0}^{N}s(k)}{N}$ - среднее значение сигнала

$E = \sum_{k=0}^{N}s^2(k)$ - энергия сигнала

$P = \frac{\sum_{k=0}^{N}s^2(k)}{N}$ - средняя моность сигнала

$s(k) = C\cos(\hat{\omega}_0k)$ - дискретный гармоический сигнал

###Пункт 1.
Смоделируем дискретный прямоугольный импульсна основе дискретного единичного скачка $σ_d(k)$ из л/р №1 с выводом графика на интервале времени $n ∈ [0,N−1]$.

```python
import numpy as np
import matplotlib.pyplot as plt


Nb = 6
N = 30+Nb%5
T = 0.0005*(1+Nb)%3
a = (-1.0)**Nb*(0.8+0.005*(Nb%20))
C = 1+Nb%5
w0 = np.pi/(6+6%5)
U = (Nb % 20) +1
n0 = 3 + (Nb % 5)
n_imp = 5 + (Nb % 5)

B1 = 1.5 + (Nb % 5)
B2 = 5.7 - (Nb % 5)
B3 = 2.2 + (Nb % 5)

w1 = np.pi / (4 + (Nb % 5))
w2 = np.pi / (8 + (Nb % 5))
w3 = np.pi / (16 + (Nb % 5))

a1 = 1.5 - (Nb % 5)
a2 = 0.7 + (Nb % 5)
a3 = 1.4 + (Nb % 5)

# Период последовательности s4
Ts4 = 2 * n_imp

# Точки интервала дискретного нормированного времени
Xn = [i for i in range(N)]
# Точки [0; 5N - 1]
Xn5 = [i for i in range(N * 5)]
# Точки 5-ти периодов последовательности s4
Xs4 = [i for i in range(Ts4 * 5)]

print(f"""
Nb = {Nb} - номер бригады
N = {N} - длина последовательности
a = {a} - основание экспоненты
C = {C} - амплитуда гармонического сигнала
w0 = {w0} - частота гармонического сигнала
U = {U} - амплитуда импульса
n0 = {n0} - начальный момент импульса
n_imp = {n_imp} - длина импульса
B1 = {B1} - амплитуда гармонического сигнала 1
B2 = {B2} - амплитуда гармонического сигнала 2
B3 = {B3} - амплитуда гармонического сигнала 3
w1 = {w1} - частота гармонического сигнала 1
w2 = {w2} - частота гармонического сигнала 2
w3 = {w3} - частота гармонического сигнала 3
a1 = {a1} - коэф. линейной комбинации гармонического сигнала 1
a2 = {a2} - коэф. линейной комбинации гармонического сигнала 2
a3 = {a3} - коэф. линейной комбинации гармонического сигнала 3
""")
```

```python
interval_norm = np.linspace(0, N-1, N)
interval_norm5 = np.linspace(0, 5*N - 1, 5*N)
# Дискретный единичный скачок
sigma_d = lambda k: np.where(k >= 0, 1, 0)
# Дискретный прямоугольный импульс на основе дискретного единичного скачка
s1 = lambda k: U * (sigma_d(k - n0) - sigma_d(k - (n0 + n_imp)))
```

```python
fig, ax = plt.subplots()
ax.stem(interval_norm, sigma_d(interval_norm), label='$\sigma_d(k)$', linefmt='g-')
ax.set_xlabel('k')
ax.set_title("Дискретный единичный скачок на интервале [0, N-1]")
ax.legend()
plt.show()

fig, ax = plt.subplots()
ax.stem(interval_norm, s1(interval_norm), label='$s_1(k)$', linefmt='g-')
ax.set_xlabel('k')
ax.set_title("Дискретный прямоугольный импульс на интервале [0, N-1]")
ax.legend()
plt.show()
```

Импульс моделируется как разность двух дискретных единичных скачков с зарежками с различием в длину импульса:

$s_1(k) = U \cdot (\sigma_d(k - n_0) - \sigma_d(k - (n_0 + n_{imp}))$

### Пункт 2.
Смоделируем линейную комбинацию дискретных гармонических сигналов $s_2(k) = a_1x_1(k) + a_2x_2(k) + a_3x_3(k)$, где $x_i(k) = B_i\sin(\hat\omega_ik), \quad i = 1, 2, 3$ с выводом графиков последовательностей $x_i(k)$ и $s_2(k)$ на интервале времени $n \in [0,5N−1]$. Вычислим среднее значение (идентификатор mean_s2), энергию (идентификатор E) и среднюю мощность последовательности
$s_2(k)$ (идентификатор P).

```python
# Гармонические сигналы
x1 = lambda k: B1 * np.sin(w1*k)
x2 = lambda k: B2 * np.sin(w2*k)
x3 = lambda k: B3 * np.sin(w3*k)

# Их линейная комбинация
s2 = lambda k: a1 * x1(k) + a2 * x2(k) + a3 * x3(k)
```

```python
fig, ax = plt.subplots()
ax.stem(interval_norm5, x1(interval_norm5), label='$x_1(k)$', linefmt='g-')
ax.set_xlabel('k')
ax.set_title("Дискретный гармонический сигнал на интервале [0, 5N-1]")
ax.legend()
plt.show()

fig, ax = plt.subplots()
ax.stem(interval_norm5, x2(interval_norm5), label='$x_2(k)$', linefmt='g-')
ax.set_xlabel('k')
ax.set_title("Дискретный гармонический сигнал на интервале [0, 5N-1]")
ax.legend()
plt.show()

fig, ax = plt.subplots()
ax.stem(interval_norm5, x3(interval_norm5), label='$x_3(k)$', linefmt='g-')
ax.set_xlabel('k')
ax.set_title("Дискретный гармонический сигнал на интервале [0, 5N-1]")
ax.legend()
plt.show()

fig, ax = plt.subplots()
ax.stem(interval_norm5, s2(interval_norm5), label='$s_2(k)$', linefmt='g-')
ax.set_xlabel('k')
ax.set_title("Линейная комбинация гармонических сигналов на интервале [0, 5N-1]")
ax.legend()
plt.show()
```

```python
s2_period = 2 * 5*9*17 # 2 * НОД знаменателей у w_i
interval_s2_period = np.linspace(0, s2_period-1, s2_period)
mean_s2 = np.mean(s2(interval_s2_period))
E = np.sum(s2(interval_s2_period) ** 2)
P = E / len(interval_s2_period)

print(f'mean_s2 = {mean_s2} , E = {E} , P = {P}')
```

1) Операции при моделировании линейной комбинации сигналов $S_2(k)$ - алгебраическая сумма произведенией гармонических сигналов на линейные коэффициенты.

$s_2(k) = a_1x_1(k) + a_2x_2(k) + a_3x_3(k)$ - линейная комбинация дискретных гармонических сигналов, где $x_i(k) = B_i\sin(\hat\omega_ik), \quad i = 1, 2, 3$

2) Определение характеристик:

$E = \sum_{k=0}^{N}s^2(k)$ - энергия сигнала

$P = \frac{\sum_{k=0}^{N}s^2(k)}{N}$ - средняя моность сигнала

### Пункт 3.
Выведем график дискретного сигнала $s_3(k)$ представляющего собой дискретный гармонический сигнал с экспоненциальной огибающей $|a|^k$, на интервале времени
$n \in [0,N−1]$.

```python
s3_exp = lambda k: np.abs(a) ** k
s3 = lambda k: C * np.cos(w0 * k) * s3_exp(k)
```

```python
fig, ax = plt.subplots()
ax.stem(interval_norm, s3_exp(interval_norm), label='$|a|^k$', linefmt='g-')
ax.set_xlabel('k')
ax.set_title("Экспоненциальная огибающая на [0, N-1]")
ax.legend()
plt.show()

fig, ax = plt.subplots()
ax.stem(interval_norm, s3(interval_norm), label='$s_3(k)$', linefmt='g-')
ax.set_xlabel('k')
ax.set_title("Дискретный сигнал с экспоненциальной огибающей на интервале [0, N-1]")
ax.legend()
plt.show()
```

1) Формула дискретного гармонического сигнала:

$s(k)=C\cos(\hat{\omega}_0k)$

2) Моделирование огибания сигнала происходит путем домножения на $|a|^k$:

$s_3(k)=C\cos(\hat{\omega}_0k)\cdot |a|^k$


### Пункт 4.
Вывести график пяти периодов периодической последовательности $s_4(k)$ дискретных прямоугольных импульсов амплитуды $U$ и длительности $n_{imp}$ c периодом, вдвое большим длительности импульса.
