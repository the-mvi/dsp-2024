## Основные теоретические положения

- Дискретный единичный скачок:

$$\sigma_d(k) = 
\begin{cases} 
1, & k \geq 0 \\ 
0, & k < 0 
\end{cases}, \; k \in \mathbb{Z}$$

- Дискретный прямоугольный импульс: 
$$s(k) = 
\begin{cases}
U, & n_0 \leq k \leq n_0 + n_{imp} - 1; \\ 
0, & \text{else}
\end{cases} $$
где $U$ - амплитуда импульса, $ n_0 $ - начальный момент импульса, $ n_{\text{imp}} $ - длина импульса.


- Среднее значение сигнала: 

$$ mean\_ s = \frac{1}{T}\sum_{n=0}^{T-1} s(k) $$
где $ T $ - период дискретного сигнала $ s(k) $

- Энергия сигнала:

$$ E = \sum_{n=0}^{T-1} s^2(k) $$
где $ T $ - период дискретного сигнала $ s(k) $

- Средняя мощность сигнала: 

$$ P = \frac{1}{T} \sum_{n=0}^{T-1} s^2(k), $$
где $ T $ - период дискретного сигнала $ s(k) $

- Дискретный гармонический сигнал с экспоненциальной огибающей $|a|^k$:

$$ s(k) = C \cos(\overset{\wedge}{\omega}_0 k)|a|^k ,$$ 
где $a$ - основание экспоненты, $C$ - амплитуда гармонического сигнала, $ \overset{\wedge}{\omega}_0$ - частота гармонического сигнала.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 32 # длина последовательности
a = -0.835 # основание экспоненты
C = 3 # амплитуда гармонического сигнала
w_0 = np.pi / 8 # частота гармонического сигнала
U = 8 # амплитуда импульса
n_0 = 5 # начальный момент импульса
n_imp = 7 # длина импульса

# амлпитуды гармонических сигналов
B_i = np.array([3.5, 3.7, 4.2])

# частоты гармонических сигналов
w_i = np.array([np.pi / 6, np.pi / 10, np.pi / 18])

# коэффициенты линейной комбинации гармонических сигналов
a_i = np.array([-0.5, 2.7, 3.4])

# моменты времени
k_1 = k_3 = np.arange(0, N) 
k_2 = np.arange(0, 5 * N)
```

### Задание 1

Для моделирования дискретного прямоугольного импульса нужно написать функцию, которая будет формировать множество значений $s_1(k)$, которое будет отображать $k$ в $U$, если $n_0 \leq k \leq n_0 + n_{imp} - 1$, иначе будет отображать в 0. Моделирование данного импульса будет подобно моделированию дискретного единичного скачка $\sigma_d(k)$, будет отличаться условие, при котором значение импульса будет ненулевым, и само значение импульса при таком условии. 

Для того, чтобы смоделировать дискретный прямоугольный импульс $s_1(k)$ на основе дискретного единичного скачка $\sigma_d(k)$, можно смоделировать разность двух смещенных дискретных единичных скачков со следующими значениями смещений: $n_0, n_0 + n_{imp}$, а затем домножить на $U$. Таким образом, итоговая формула будет выглядеть так: $$s_1(k) = U (\sigma_d(k - n_0) - \sigma_d(k - (n_0 + n_{imp})))$$

```python
def plot_for_signals(x, y, label): 
    plt.figure(figsize=(12, 5))
    plt.scatter(x, y, color='purple')
    for xi, yi in zip(x, y):
        plt.plot([xi, xi], [0,yi], color='purple')
    plt.title(label)
    plt.grid(True)
    plt.show()
    
def model_discreate_unit_leap(time_array):
    result = np.zeros_like(time_array)
    positive_indices = time_array >= 0
    result[positive_indices] = 1
    return result
    
def model_discreate_rectangle_impulse_from_unit_leap(time_array):
    rectangle_impulse = U * (model_discreate_unit_leap(time_array - n_0) - model_discreate_unit_leap(time_array - n_0 - n_imp))
    return rectangle_impulse

plot_for_signals(k_1, model_discreate_rectangle_impulse_from_unit_leap(k_1), "График дискретного прямоугольного импульса $s_1(k)$")
```

### Задание 2

a. Для моделирования линейной комбинации сигналов $s_2(k)$ нужно:
- Вычислить каждый дискретный гармонический сигнал $x_i(k)$ на интервале времени $ n \in [0,\ 5N - 1]$ с помощью формулы: $$ x_i(k) = B_i\sin(\overset{\wedge}{\omega}_i * k) $$
- Вычислить сумму этих сигналов: $$ s_2(k) = \sum_{i = 1}^{n} a_i x_i(k), $$  
где $a_i$ - коэффициенты линейной комбинации гармонических сигналов


```python
def calculate_term(B_i, w_i, time_array):
    x_i = np.zeros_like(time_array)
    x_i = B_i * np.sin(w_i * time_array)
    return x_i

def model_linear_combination(time_array):
    s_2 = np.zeros_like(time_array)
    s_2 = a_i[0] * calculate_term(B_i[0], w_i[0], time_array) + a_i[1] * calculate_term(B_i[1], w_i[1], time_array) + a_i[2] * calculate_term(B_i[2], w_i[2], time_array)
    return s_2

plot_for_signals(k_2, model_linear_combination(k_2), "График линейной комбинации дискретных гармонических сигналов $s_2(k)$")
for i in range(3):
    graphic_label = "График дискретного гармонического сигнала $ x_" + str(i + 1) + "(k)$"
    plot_for_signals(k_2, calculate_term(B_i[i], w_i[i], k_2), graphic_label)
```

b. 
- Среднее значение сигнала (идентификатор mean_s2): $$ mean\_ s2 = \frac{1}{T}\sum_{n=0}^{T-1} s_2(n) $$
- Энергия сигнала (идентификатор E): $$ E = \sum_{n=0}^{T-1} s_2^2(n) $$
- Средняя мощность сигнала (идентификатор P): $$ P = \frac{1}{T} \sum_{n=0}^{T-1} s_2^2(n) $$

Как можно заметить из графика линейной комбинации дискретных гармонических сигналов, период T явно больше, чем $5N - 1$, поэтому для вычисления таких численных характеристик, как среднее значение, энергия и средняя мощность, нужно найти период данной последовательности. 

Ниже представлен код для нахождения периода последовательности $s_2(k)$.

```python
def find_period(signal):
    n = len(signal)
    for period in range(1, n // 2 + 1):
        if all(signal[i] - signal[i + period] < 1e-10 for i in range(n - period)):
            return period
    return None 

period = find_period(model_linear_combination(np.arange(0, 20 * N + 1)))
print("Период сигнала:", period)
```

Был найден период последовательности $s_2(k)$, он равен  180. Теперь можно посчитать необходимые характеристики.

```python
def find_mean_value(signal):
    return np.mean(signal)

def find_energy(signal):
    return np.sum(signal ** 2)

def find_mean_power(signal):
    return np.mean(signal ** 2)

signal = model_linear_combination(np.arange(0, period))
mean_s2 = find_mean_value(signal)
E = find_energy(signal)
P = find_mean_power(signal)

print("Среднее значение:", mean_s2)
print("Энергия:", E)
print("Средняя мощность последовательности", P) 
```

### Задание 3

Аналитическая формула дискретного сигнала $s_3(k)$: 
$$ s_3(k) = C \cos(\overset{\wedge}{\omega}_0 k)|a|^k $$ 

Для моделирования дискретного гармонического сигнала с экспоненциальной огибающей на интервале времени $ n \in [0, N - 1] $ нужно:
- Вычислить значения дискретного гармонического сигнала $s(k)$:
$$ s(k) = C \cos(\overset{\wedge}{\omega}_0 k) $$ 
- Вычислить значения экспоненциальнной огибающей $|a|^k$
- Вычислить произведение : $$ s_3(k) = s(k) |a|^k $$

Экспоненциальная огибающая позволяет контролировать амплитуду сигнала с течением времени, создавая эффект затухания или нарастания.

```python
def model_discreate_exponential_harmonic_signal(time_array):
    s_i = C * np.cos(w_0 * time_array)
    s_3 = s_i * np.abs(a)**time_array
    return s_3

def plot_for_exp(x, y, label): 
    plt.figure(figsize=(12, 5))
    plt.scatter(x, y, color='purple')
    for xi, yi in zip(x, y):
        plt.plot([xi, xi], [0,yi], color='purple')
    plt.plot(x, C*np.abs(a)**x, color = "red", linestyle='dashed')
    plt.plot(x, C*(-np.abs(a)**x), color = "red", linestyle='dashed')
    plt.title(label)
    plt.grid(True)
    plt.show()
    

plot_for_exp(k_3, model_discreate_exponential_harmonic_signal(k_3), "График дискретного гармонического сигнала с экспоненциальной огибающей $s_3(k)$")
```

### Задание 4

Для моделирования периодической последовательности нужно:
- Смоделировать 5 прямоугольных импульсов длины $n_{imp}$ и амплитуды $U$, расстояние между которыми будет равным $n_{imp}$(так как период должен быть вдвое больше длительности импульса).

Таким образом, 1 период в таком графике будет представлять следующее: прямоугольный импульс длины $n_{imp}$ и амплитуды $U$, затем нулевое значение для оси $y$ при последующих $n_{imp}$ значениях аргумента.

```python
time_array = np.arange(0, 5 * (n_imp * 2))

def model_five_discreate_rectangle_impulses(time_array):
    result = np.zeros_like(time_array)
    for i in range(5):
        result[i * (n_imp * 2): i * (n_imp * 2) + n_imp] = U  
    return result

plot_for_signals(time_array, model_five_discreate_rectangle_impulses(time_array), "График пяти периодов последовательности $s_4(k)$ дискретных прямоугольных импульсов")
```
