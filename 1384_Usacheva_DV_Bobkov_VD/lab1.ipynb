## Основные теоретические положения

Аналоговый сигнал $s(t)$ — сигнал, у которого каждый из представленных параметров описывается функцией времени, которая может принимать любое значение в непрерывном диапазоне.

Дискретные сигналы $s(nT)$, $s(n)$ — это последовательность значений, изменяющихся в дискретные моменты времени, где $nT$ — дискретное время, а $n$ — дискретное нормированное время.

Единичный цифровой импульс:

$
\delta_d(k) = 
\begin{cases} 
1, & k = 0 \\ 
0, & k \neq 0 
\end{cases}, \; k \in \mathbb{Z}
$

Является дискретным аналогом дельта-функции:

$
\delta(t) =
\begin{cases}
\infty, & t = 0 \\
0, & t \neq 0
\end{cases}, \quad \int_{-\infty}^{+\infty} \delta(t) \, dt = 1
$

Дискретный единичный скачок:

$
\sigma_d(k) = 
\begin{cases} 
1, & k \geq 0 \\ 
0, & k < 0 
\end{cases}, \; k \in \mathbb{Z}
$

Является дискретным аналогом функции единичного скачка:

$
\sigma(t) = 
\begin{cases} 
1, & t > 0 \\ 
0.5 \text{ или не определена}, & t = 0 \\ 
0, & t < 0 
\end{cases}
$


Частота дискретизации:

$
f_d = \frac{1}{T},$ где $T$ — период дискретизации.

Дискретная экспоненциальная функция:

$
s_1(k) = 
\begin{cases} 
a^k, & k \geq 0 \\ 
0, & k < 0 
\end{cases}, \;  k \in \mathbb{Z}
$

Дискретный комплексный гармонический сигнал:

$
s_2(k) = C \exp(j \overset{\wedge}{\omega}_0 k)
$

## Выполнение работы

```python
import numpy as np
import matplotlib.pyplot as plt

N = 32 # длина последовательности
T = 0.001 # период дискретизации
a = -0.835 # основание экспоненты
C = 3 # амплитуда гармонического сигнала
w0 = np.pi/8 # частота гармонического сигнала
m = 7 # задержка 

n = np.arange(0, N) # дискретное нормированное время
nt = n * T # дискретное время
```

### Задание 1

```python
def model_unit_impulse(time_array): # функция для моделирования единичного импульса
    result = np.zeros_like(time_array)
    result[time_array == 0] = 1
    return result

def plot_for_signals(x, y, label): # функция для построения графика
    plt.scatter(x, y, color='pink')
    for xi, yi in zip(x, y):
        plt.plot([xi, xi], [0,yi], color='pink')
    plt.title(label)
    plt.grid(True)
    plt.show()
    
delta_d = model_unit_impulse(n)
plot_for_signals(nt, delta_d, "График единичного цифрового импульса на интервале дискретного времени")
plot_for_signals(n, delta_d, "График единичного цифрового импульса на интервале дискретного нормированного времени")
```

Были построены графики единичного цифрового импульса на интервале дискретного времени и дисретного нормированного времени.

a. Была получена взаимосвязь между дискретным и дискретным нормированным временем: номер отсчета $n$ является значением нормированного времени, то есть номер $n$ означает, что отсчет взят в момент $nT$, где T - частота дискретизации. Можно установить соотношение этих величин:

$t$ - дискретное время, $\overset{\wedge}{t}$ - дискретное нормированное время, $T$ - период дискретизации, тогда $\overset{\wedge}{t} = \frac{t}{T}$

b. Единичный цифровой импульс является дискретным аналогом дельта-функции (функции Дирака). Различие между единичным цифровым импульсом и дельта-функцией заключается в том, что единичный цифровой импульс является дискретным сигналом и принимает значение 1 в точке $t = 0$, $0$ в остальных случаях. Дельта-функция (функции Дирака) является аналоговым сигналом и имееет бесконечное значение в точке $t = 0$, $0$ в остальных случаях.



### Задание 2

```python
def model_discreate_unit_leap(time_array):
    result = np.zeros_like(time_array)
    positive_indices = time_array >= 0
    result[positive_indices] = 1
    
    return result

sigma_d = model_discreate_unit_leap(n)
plot_for_signals(nt, sigma_d, "График дискретного цифрового скачка на интервале дискретного времени")
plot_for_signals(n, sigma_d, "График дискретного цифрового скачка на интервале дискретного нормированного времени")

```

Были построены графики единичного цифрового скачка на интервале дискретного времени и дисретного нормированного времени.

a. Дискретный единичный скачок $\sigma_d(k)$ является дискретным аналогом функции единичного скачка $\sigma(t)$ (функции Хевисайда). 

b. Частоту дискретизации $f_d$ можно получить по формуле $f_d = \frac{1}{T} = 1 кГц$


### Задание 3

```python
def model_discreate_exponential(time_array): 
    result = np.zeros_like(time_array, dtype=float)
    positive_indices = time_array >= 0
    result[positive_indices] = a ** time_array[positive_indices]
    return result
    
s_1 = model_discreate_exponential(n)
plot_for_signals(nt, s_1, "График дискретной экспоненциальной функции на интервале дискретного времени")
plot_for_signals(n, s_1, "График дискретной экспоненциальной функции на интервале дискретного нормированного времени")
```

Дискретная экспонента может быть получена из аналоговой экспоненты путем дискретизации, то есть путем выборки значений аналоговой экспоненты в определенные моменты времени.

### Задание 4

С помощью формулы Эйлера можно расписать данный сигнал в виде двух вещественных последовательностей: $C * e^{j\overset{\wedge}{\omega}_0k} = C * \cos(\overset{\wedge}{\omega}_0k) + C * j \sin(\overset{\wedge}{\omega}_0k)$. Таким образом получается следующее: 

1-ая последовательность — $C * \cos(\overset{\wedge}{\omega}_0k)$, 

2-ая последовательность — $C * \sin(\overset{\wedge}{\omega}_0k)$.

```python
def model_discreate_complex_harmonic(time_array):
    real_part = np.zeros_like(time_array)
    complex_part = np.zeros_like(time_array)
    positive_indices = time_array >= 0
    real_part[positive_indices] = C * np.cos(w0 * time_array[positive_indices])
    complex_part[positive_indices] = C * np.sin(w0 * time_array[positive_indices])
    return real_part, complex_part

discreat_complex_harmonic = model_discreate_complex_harmonic(n)
plot_for_signals(n, discreat_complex_harmonic[0],"График вещественной части дискретного комплекного гармонического сигнала")
plot_for_signals(n, discreat_complex_harmonic[1],"График мнимой части дискретного комплекного гармонического сигнала")
```

Были построены графики вещественной и мнимой части дискретного комплекного гармонического сигнала, также с помощью формулы Эйлера данный сигнал был записан в виду комбинации двух вещественных последовательностей:  $C * e^{j\overset{\wedge}{\omega}_0k} = C * \cos(\overset{\wedge}{\omega}_0k) + C * j \sin(\overset{\wedge}{\omega}_0k)$.
