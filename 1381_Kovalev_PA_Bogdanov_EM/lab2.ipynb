### Задание 1

Смоделировать дискретный прямоугольный импульс:$$ s_1(k) = \begin{cases} U,\quad n_0 \leqslant k \leqslant n_0+n_\text{imp}-1 \\ 0, \quad else \end{cases} $$
на основе дискретного единичного скачка $$σ_d(k)$$ из л/р №1 с выводом графика на интервале времени $$n∈[0,N−1]$$ Пояснить как выполняется моделирование импульса.


```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0, N)

# Определение дискретного единичного скачка σd(k)
def sigma_d(k):
    return np.where(k >= 0, 1, 0)

# Моделирование дискретного прямоугольного импульса s1(k)
s1 = U * (sigma_d(n - n0) - sigma_d(n - (n0 + nimp)))

plt.stem(n, s1)
plt.title('Дискретный прямоугольный импульс $s_1(k)$')
plt.xlabel('n')
plt.ylabel('s1(k)')
plt.grid(True)
plt.show()
```

Моделирование дискретного прямоугольного импульса $ s_1(k) $:

Задача состоит в том, чтобы создать сигнал $ s_1(k) $, который имеет значение амплитуды $U$ в интервале от $n_0$ до $n_0 + n_{\text{imp}} - 1$ и нулевое значение вне этого интервала.

Импульс определяется как:

$$
s_1(k) = U \cdot [\sigma_d(k - n_0) - \sigma_d(k - (n_0 + n_{\text{imp}}))]
$$

Разность единичных скачков: 

При вычитании второго скачка из первого, получается прямоугольный импульс, который равен 1 в интервале $ n_0 \leq k < n_0 + n_{\text{imp}} $ и $0$ вне этого интервала.


### Задание 2
Смоделировать линейную комбинацию дискретных гармонических сигналов $s_2(k)$

с выводом графиков последовательностей $x_i(k)$ и $s_2(k)$ на интервале времени $n∈[0,5N−1]$ Вычислить среднее значение (идентификатор mean_s2), энергию (идентификатор E) и среднюю мощность (идентификатор P). 

Пояснить:

a. операции при моделировании линейной комбинации сигналов;

b. как определяют указанные характеристики.


```python
import numpy as np
import matplotlib.pyplot as plt
import math as Math

Nb = 5
N = 30 + (Nb % 5)
a = (-1)**Nb * (0.8 + 0.005 * (Nb % 20))
C = 1 + (Nb % 5)
omega0 = np.pi / (6 + (Nb % 5))
U = (Nb % 20) + 1
n0 = 3 + (Nb % 5)
n_imp = 5 + (Nb % 5)

B1 = 1.5 + (Nb % 5)
B2 = 5.7 - (Nb % 5)
B3 = 2.2 + (Nb % 5)

omega1 = np.pi / (4 + (Nb % 5))
omega2 = np.pi / (8 + (Nb % 5))
omega3 = np.pi / (16 + (Nb % 5))

a1 = 1.5 - (Nb % 5)
a2 = 0.7 + (Nb % 5)
a3 = 1.4 + (Nb % 5)

def x1(k):
    return B1*Math.sin(omega1*k)

def x2(k):
    return B2*Math.sin(omega2*k)

def x3(k):
    return B3*Math.sin(omega3*k)

def s_2(k):
    return a1*x1(k)+a2*x2(k)+a3*x3(k)

X1=[]
Y1=[]
X2=[]
Y2=[]
X3=[]
Y3=[]
X=[]
Y=[]
for i in range (5*N-1):
    X1.append(i)
    Y1.append(x1(i))

    X2.append(i)
    Y2.append(x2(i))

    X3.append(i)
    Y3.append(x3(i))
    
    X.append(i)
    Y.append(s_2(i))


plt.stem(X,Y)
plt.title('Линейная  комбинация  дискретных  гармонических  сигналов  $s_2(k)$')
plt.xlabel("n")
plt.ylabel(r'$s_2(k)$')
plt.show()

plt.stem(X1,Y1)
plt.title(r'$x_1$')
plt.xlabel("n")
plt.ylabel(r'$x_1(k)$')
plt.show()

plt.stem(X2,Y2)
plt.title(r'$x_2$')
plt.xlabel("n")
plt.ylabel(r'$x_2(k)$')
plt.show()

plt.stem(X3,Y3)
plt.title(r'$x_3$')
plt.xlabel("n")
plt.ylabel(r'$x_3(k)$')
plt.show()

T=32
mean_s2=0
E=0
P=0
for i in range(T):
    mean_s2+=Y[i]
    E=Y[i]**2
P=E/T
print("Mean s_2=", mean_s2)

print("E=", E)

print("P=", P)

```

a)Для получения линейной комбинации необходимо смоделировать периодические сигналы 
$$X_i=B_isin(\hat{\omega_i}k)$$

и взять их с "весами" a

b)Характеристики определятся по формулам. Для расчётов необходим период получившейся линейной комбинации. Он был расчитан как НОК периодов периодических сигналов:
$$T_1=\frac{2\pi}{\omega_1}=\frac{2\pi}{\frac{\pi}{4}}=8$$ 
$$T_2=\frac{2\pi}{\omega_2}=\frac{2\pi}{\frac{\pi}{8}}=16$$
$$T_3=\frac{2\pi}{\omega_3}=\frac{2\pi}{\frac{\pi}{16}}=32$$

$$T=LCM(T_1,T_2,T_3)=32$$


### Задание 3

Вывести график дискретного сигнала $s_3(k)$ представляющего собой дискретный гармонический сигнал $s(k)=Ccos(\hat{ω}_0k)$с экспоненциальной огибающей $|a|^k$ на интервале времени $n∈[0,N−1]$ Привести аналитическую формулу дискретного сигнала $s_3(k)$ и пояснить операции при его моделировании.

```python
#3
import numpy as np
import matplotlib.pyplot as plt
import math as Math

Nb = 5
N = 30 + (Nb % 5)
a = (-1)**Nb * (0.8 + 0.005 * (Nb % 20))
C = 1 + (Nb % 5)
omega0 = np.pi / (6 + (Nb % 5))
U = (Nb % 20) + 1
n0 = 3 + (Nb % 5)
n_imp = 5 + (Nb % 5)

B1 = 1.5 + (Nb % 5)
B2 = 5.7 - (Nb % 5)
B3 = 2.2 + (Nb % 5)

omega1 = np.pi / (4 + (Nb % 5))
omega2 = np.pi / (8 + (Nb % 5))
omega3 = np.pi / (16 + (Nb % 5))

a1 = 1.5 - (Nb % 5)
a2 = 0.7 + (Nb % 5)
a3 = 1.4 + (Nb % 5)

def s(k):
    return C*Math.cos(omega0*k)

def e(k):
    return abs(a)**k

def s3(k):
    return e(k)*s(k)

X=[]
Y=[]
Y1=[]
Y2=[]
for i in range(N-1):
    X.append(i)
    Y.append(s3(i))
    Y1.append(e(i))
    Y2.append(-1*e(i))

plt.title('Дискретный гармонический сигнал  $s_3(k)$')
plt.stem(X,Y)
plt.plot(X,Y1,color="green")
plt.plot(X,Y2,color="green")
plt.xlabel("n")
plt.ylabel(r'$s_3(k)$')

plt.show()


```

Аналитическая формула:
$$s_3(k)=|a|^k*C*cos(\hat{ω}_0k)$$

Экспонента определяет амплитуду гармонического сигнала. Её основание меньше единицы,поэтому амплитуда стремится к нулю. Для получение сигнала,необходмо умножить $s(k)$ на амплитуду $|a|^k*C$

### Задание 4

Вывести график пяти периодов периодической последовательности $s_4(k)$ дискретных прямоугольных импульсов амплитуды U и длительности nimp с периодом, вдвое большим длительности импульса. Пояснить операции при моделировании периодической последовательности.


```python
import numpy as np
import matplotlib.pyplot as plt

Nb = 5

N = 30 + (Nb % 5)

a = (-1)**Nb * (0.8 + 0.005 * (Nb % 20))

C = 1 + (Nb % 5)

omega0 = np.pi / (6 + (Nb % 5))

U = (Nb % 20) + 1

n0 = 3 + (Nb % 5)

nimp = 5 + (Nb % 5)

B1 = 1.5 + (Nb % 5)

B2 = 5.7 - (Nb % 5)

B3 = 2.2 + (Nb % 5)

omega1 = np.pi / (4 + (Nb % 5))

omega2 = np.pi / (8 + (Nb % 5))

omega3 = np.pi / (16 + (Nb % 5))

a1 = 1.5 - (Nb % 5)

a2 = 0.7 + (Nb % 5)

a3 = 1.4 + (Nb % 5)


# Параметры периодической последовательности
T = 2 * nimp                  # Период последовательности (вдвое больше длительности импульса)
num_periods = 5               # Количество периодов для отображения
N = num_periods * T           # Общее количество отсчетов

# Создание последовательности n
n = np.arange(0, N)

# Моделирование периодической последовательности s4(k)
m = n % T                     # Значения внутри одного периода
s4 = np.where(m < nimp, U, 0) # Формирование импульса в каждом периоде

plt.figure(figsize=(12, 4))
plt.stem(n, s4)
plt.title('Периодическая последовательность s4(k)')
plt.xlabel('n')
plt.ylabel('s4(k)')
plt.grid(True)
plt.show()
```

Для создания периодичности используем операцию взятия остатка от деления на период $T$. Для каждого значения $n$ вычисляется соответствующее значение $m$, которое повторяется от $0$ до $T - 1$ в каждом периоде.

Если текущая позиция внутри периода $ m $ меньше длительности импульса $ n_{\text{imp}} $, то в этом месте сигнал принимает значение $U$.
