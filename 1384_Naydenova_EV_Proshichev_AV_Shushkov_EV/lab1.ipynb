## Основные теоретические положения
Цифровые сигналы представляют собой дискретные формы сигналов, которые изменяются через определенные промежутки времени и могут быть представлены последовательностью значений. 

Аналоговые сигналы $s(t)$, в отличие от цифровых, являются непрерывными и могут принимать любое значение в пределах заданного диапазона. 

Дискретные сигналы $s(nT)$ формируются путем выборки аналогового сигнала в определенные моменты времени. 

Частота дискретизации указывает на то, с какой частотой аналоговый сигнал преобразуется в дискретный.

Взаимосвязь между дискретным и дискретным нормированным временем заключается в том, что дискретное время изменяется через равные интервалы с целочисленным индексом $ n $ и периодом дискретизации $ T $. Дискретное нормированное время представляет собой преобразованную версию дискретного времени, учитывающую относительные интервалы времени.

Единичный цифровой импульс $ \delta_d(k) $ используется для дискретных сигналов и определяется следующим образом $:

$
\delta_d(k) = 
\begin{cases} 
1, & k = 0 \\ 
0, & k \neq 0 
\end{cases}, \; k \in \mathbb{Z}
$

Дельта-функция $ \delta(t) $ предназначена для непрерывных сигналов :

$
\delta(t) = 
\begin{cases} 
\infty, & t = 0 \\ 
0, & t \neq 0 
\end{cases}, \; t \in \mathbb{R}
$

Это условие подразумевает, что интеграл дельта-функции равен единице.

Дискретный скачок $ \sigma_d(k) $ представляется следующим образом:

$
\sigma_d(k) = 
\begin{cases} 
1, & k \geq 0 \\ 
0, & k < 0 
\end{cases}, \; k \in \mathbb{Z}
$

Аналоговый скачок $ \sigma(t) $ определяется следующим образом:

$
\sigma(t) = 
\begin{cases} 
1, & t > 0 \\ 
0.5 \text{ или не определена}, & t = 0 \\ 
0, & t < 0 
\end{cases}, \;  t \in \mathbb{R}
$

Частота дискретизации $ f_d $ определяется как обратная величина периода дискретизации $ T $:

$
f_d = \frac{1}{T}
$

Частота дискретизации позволяет понять, с какой частотой аналоговый сигнал преобразуется в дискретный.

Дискретная экспонента определяется следующим образом:

$
s_1(k) = 
\begin{cases} 
a^k, & k \geq 0 \\ 
0, & k < 0 
\end{cases}, \;  k \in \mathbb{Z}
$

Аналоговая экспонента задается как:

$
s(t) = 
\begin{cases} 
a^t, & t \geq 0 \\ 
0, & t < 0 
\end{cases}, \;  t \in \mathbb{R} 
$

Дискретный комплексный гармонический сигнал $ s_2(k) $ задается следующим образом, где $ k \in \mathbb{Z}$:

$
s_2(k) = C \exp(j \hat{\omega_0} k)
$

Этот сигнал можно записать в виде комбинации двух вещественных последовательностей с использованием формулы Эйлера:

$
s_2(k) = C \cdot (\cos(\hat{\omega_0} k) + j \sin(\hat{\omega_0} k))
$

## Выполнение работы

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Константы
N_br = 10  # Номер бригады
N = 30 + N_br % 5  # Длина последовательности
T = 0.0005 * (1 + N_br % 3)  # Период дискретизации
a = (-1) ** N_br * (0.8 + 0.005 * (N_br % 20))  # Основание экспоненты
C = 1 + N_br % 5  # Амплитуда гармонического сигнала
omega_0 = math.pi / (6 + N_br % 5)  # Частота гармонического сигнала
m = 5 + N_br % 5  # Задержка

# Вывод констант
print(f"N_br: {N_br}")
print(f"N: {N}")
print(f"T: {T}")
print(f"a: {a}")
print(f"C: {C}")
print(f"omega_0: {omega_0}")
print(f"m: {m}")
```

### Задание 1

Смоделировать единичный цифровой импульс $ \delta_d(k) $ с выводом графиков на интервале дискретного времени $ nT \in [0, (N-1)T] $ и дискретного нормированного времени $ n \in [0, N-1] $. Пояснить: <br>
 - Взаимосвязь между дискретным и дискретным нормированным временем <br>
 - Различие между единичным цифровым импульсом и дельта-функцией


```python
# Дискретное время
n = np.arange(0, N)
nT = n * T

# Единичный цифровой импульс
delta_d = np.zeros(N)
delta_d[0] = 1  # Импульс

# График
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.stem(nT, delta_d, linefmt="b-", markerfmt="bo")
plt.title('Единичный цифровой импульс δ_d(k)')
plt.ylabel('Амплитуда')

plt.subplot(2, 1, 2)
plt.stem(n, delta_d, linefmt="b-", markerfmt="bo")
plt.title('Единичный цифровой импульс δ_d(k) (нормированное время)')
plt.ylabel('Амплитуда')

plt.tight_layout()
plt.show()
```

a. Взаимосвязь между дискретным и дискретным нормированным временем <br>
Дискретное время $nT$ — это время, изменяющееся через равные интервалы, где $ n $ — целочисленный индекс, а $ T $ — период дискретизации. 

Дискретное нормированное время $n$ — это время, также изменяющееся дискретно, но с учетом нормировки. 

Взаимосвязь между ними заключается в том, что дискретное нормированное время является преобразованной версией дискретного времени, учитывающей относительные интервалы времени.

b. Различие между единичным цифровым импульсом и дельта-функцией <br>

Единичный цифровой импульс $ \delta_d(k) $ используется для дискретных сигналов.

$\delta_d(k) = \begin{cases}
1, &  k = 0 \\
0, &  k \neq 0
\end{cases}$,  $\; k \in \mathbb{Z}$


Дельта-функция $ \delta(t) $ используется для непрерывных сигналов .

$ \delta(t) = \begin{cases}
\infty, &  t = 0 \\
0, &  t \neq 0
\end{cases}  $,  $ \; t \in \mathbb{R}$

с условием:

$ \int_{-\infty}^{\infty} \delta(t) \, dt = 1 $

### Задание 2
Смоделировать дискретный единичный скачок $ \sigma_d(k) $ с выводом графиков на интервале дискретного времени $ nT \in [0, (N-1)T] $ и дискретного нормированного времени $ n \in [0, N-1] $. Пояснить:
- Соответствие между дискретным и аналоговым единичными скачками.
- Чему равна частота дискретизации дискретного единичного скачка.


```python
# Дискретный единичный скачок
sigma_d = np.zeros(N)
sigma_d[0:] = 1  

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.stem(nT, sigma_d, linefmt="b-", markerfmt="bo")
plt.title('Дискретный единичный скачок σ_d(k)')
plt.ylabel('Амплитуда')
plt.grid(True, linestyle='--', alpha=0.7)

plt.subplot(2, 1, 2)
plt.stem(n, sigma_d, linefmt="b-", markerfmt="bo")
plt.title('Дискретный единичный скачок σ_d(k) (нормированное время)')
plt.ylabel('Амплитуда')

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

a. Соответствие между дискретным и аналоговым единичными скачками

Скачок $ \sigma_d(k) $ - является дискретным аналогом функции единичного скачка (Хэвисайда)

$\sigma_d(k) = \begin{cases}
1, &  k \geq 0 \\
0, &  k < 0
\end{cases}$,   $\; k \in \mathbb{Z}$

Аналоговый единичный скачок

$\sigma(t) = \begin{cases}
1, &  t > 0 \\
0.5  \text{ или не определена}, &  t = 0 \\ 
0, &  t < 0
\end{cases}$ ,  $\; t \in \mathbb{R}$

Основные различия:

Дискретный: Время дискретно, представляется последовательностью значений. <br>
Аналоговый: Время непрерывно, представляется функцией непрерывного времени.

b. Частота дискретизации дискретного единичного скачка

Частота дискретизации $ f_d $ определяется как обратная величина периода дискретизации $ T $:

$ f_d = \frac{1}{T} = \frac{1}{0.001} = 1000 \; \text{Гц}$

где $ T $ — период дискретизации.

### Задание 3
Смоделировать дискретную экспоненциальную функцию $ s_1(k) $ с выводом графиков на интервале дискретного времени $ nT \in [0, (N-1)T] $ и дискретного нормированного времени $ n \in [0, N-1] $. Пояснить соответствие между дискретной и аналоговой экспонентами.


```python
s_1 = a ** n

plt.figure(figsize=(12, 6))

# График в дискретном времени
plt.subplot(2, 1, 1)
plt.stem(nT, s_1, linefmt="r-", markerfmt="ro")
plt.title('Дискретная экспоненциальная функция s_1(k)')
plt.ylabel('Амплитуда')
plt.grid(True, linestyle='--', alpha=0.7)

# График в нормированном времени
plt.subplot(2, 1, 2)
plt.stem(n, s_1, linefmt="r-", markerfmt="ro")
plt.title('Дискретная экспоненциальная функция s_1(k) (нормированное время)')
plt.ylabel('Амплитуда')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Соответствие между дискретной и аналоговой экспонентами

Дискретная экспонента

$s_1(k) = \begin{cases} 
a^k, & k \geq 0 \\ 
0, &  k < 0 
\end{cases}$ ,  $\; k \in \mathbb{Z}$

Аналоговая экспонента

$s(t) = \begin{cases} 
a^t, &  t \geq 0 \\ 
0, &  t < 0 
\end{cases}$ ,  $\; t \in \mathbb{R}$

Основные различия:

Дискретная экспонента: Время представлено как последовательность отсчетов, и значения функции изменяются на каждом шаге.

Аналоговая экспонента: Время непрерывно, и функция представляет собой плавное изменение, без "скачков".

### Задание 4
Смоделировать дискретный комплексный гармонический сигнал $s_2(k) = C \exp(j \hat{\omega_0} k)$ с выводом графиков вещественной и мнимой частей на интервале времени $ n \in [0, N-1] $ Записать данный сигнал в виде комбинации двух вещественных последовательностей.

```python
s_2 = C * np.exp(1j * omega_0 * n)  # Комплексный гармонический сигнал

# Вещественная и мнимая части
real_part = np.real(s_2)
imag_part = np.imag(s_2)


# Построение графиков
plt.figure(figsize=(12, 6))

# График вещественной части в дискретном времени
plt.subplot(2, 1, 1)
plt.stem(n, real_part, linefmt="b-", markerfmt="bo")
plt.title('Вещественная часть комплексного гармонического сигнала')
plt.ylabel('Амплитуда')
plt.grid(True, linestyle='--', alpha=0.7)

# График мнимой части в дискретном времени
plt.subplot(2, 1, 2)
plt.stem(n, imag_part, linefmt="r-", markerfmt="ro")
plt.title('Мнимая часть комплексного гармонического сигнала')
plt.ylabel('Амплитуда')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Сигнал можно разложить на вещественную и мнимую части с помощью формулы Эйлера:

$s_2(k) = C \cdot (\cos(\omega_0 k) + j \sin(\omega_0 k))$

Таким образом, сигнал $ s_2(k) $ можно записать как комбинацию двух вещественных последовательностей:

$s_2(k) = C \cdot \cos(\omega_0 k) + j C \cdot \sin(\omega_0 k)$

$s_2(k) = \cos(\frac{\pi k}{6}) + j  \sin(\frac{\pi k}{6})$

$s_2Re(k) = \cos(\frac{\pi k}{6})$

$s_2Im(k) = \sin(\frac{\pi k}{6}) $


### Задание 5
Вывести графики последовательностей  $\delta_d(k), \sigma_d(k), \text{и } s_1(k)$ задержанных на $m $ отсчетов, на интервале времени $n \in [0, N-1]$ Записать формулы задержанных последовательностей.

Формула задержанного единичного цифрового импульса:

$
\delta_d(k - m) = 
\begin{cases} 
1, &  k = m \\ 
0, &  k \neq m 
\end{cases}
$  ,  $\; k \in \mathbb{Z}$

где $ m $ — количество задержанных отсчетов.

```python
# Дискретное время
n = np.arange(0, N)
nT = n * T

# Единичный цифровой импульс
delta_d = np.zeros(N)
delta_d[m] = 1  # Импульс

# График
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.stem(nT, delta_d, linefmt="b-", markerfmt="bo")
plt.title('Задержанный единичный цифровой импульс δ_d(k)')
plt.ylabel('Амплитуда')

plt.subplot(2, 1, 2)
plt.stem(n, delta_d, linefmt="b-", markerfmt="bo")
plt.title('Задержанный единичный цифровой импульс δ_d(k) (нормированное время)')
plt.ylabel('Амплитуда')

plt.tight_layout()
plt.show()
```

Формула задержанного единичного скачка :

$
\sigma_d(k - m) = 
\begin{cases} 
1, &  k \geq m \\ 
0, &  k < m 
\end{cases}
$ ,  $\; k \in \mathbb{Z}$

где $ m $ — количество задержанных отсчетов.

```python
# Дискретный единичный скачок
sigma_d = np.zeros(N)
sigma_d[m:] = 1  

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.stem(nT, sigma_d, linefmt="b-", markerfmt="bo")
plt.title('Задержанный дискретный единичный скачок σ_d(k)')
plt.ylabel('Амплитуда')
plt.grid(True, linestyle='--', alpha=0.7)

plt.subplot(2, 1, 2)
plt.stem(n, sigma_d, linefmt="b-", markerfmt="bo")
plt.title('Задержанный дискретный единичный скачок σ_d(k) (нормированное время)')
plt.ylabel('Амплитуда')

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Формула задержанной экспоненты :

$
s_1(k - m) = 
\begin{cases} 
a^{k - m}, & k \geq m \\ 
0, &  k < m 
\end{cases}
$,  $\; k \in \mathbb{Z}$

где $ a $ — коэффициент затухания, $ m $ — количество задержанных отсчетов.

```python
s_1 = np.zeros(N) 
s_1[m:] = C * a ** (n[m:])  # Задержанная экспонента

plt.figure(figsize=(12, 6))

# График в дискретном времени
plt.subplot(2, 1, 1)
plt.stem(nT, s_1, linefmt="r-", markerfmt="ro")
plt.title('Дискретная экспоненциальная функция s_1(k)')
plt.ylabel('Амплитуда')
plt.grid(True, linestyle='--', alpha=0.7)

# График в нормированном времени
plt.subplot(2, 1, 2)
plt.stem(n, s_1, linefmt="r-", markerfmt="ro")
plt.title('Дискретная экспоненциальная функция s_1(k) (нормированное время)')
plt.ylabel('Амплитуда')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
