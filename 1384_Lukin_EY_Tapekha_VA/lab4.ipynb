### Задание 1
Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона. Построить графики передаточных функций и графики отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства полученных передаточных функций.

#### Квадратурная формула прямоугольников
Выведем формулу передаточной функции рекурсивного фильтра, соответствующей квадратурной формуле прямоугольников $ y_{n+1} = y_n + s_{n + \frac{1}{2}},\ (1)$ где $y_0 = 0.$ Отсчет входного сигнала как $s_n=e^{i\omega n},$ выход фильтра $y_n=H(\omega)e^{i\omega n}.$ Тогда, подставляя эти выражения в уравнение (1), получим:
$$
\begin{align*}
   & H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega n} + e^{i\omega \left(n+\frac{1}{2}\right )} \Leftrightarrow
    H(\omega)\left( e^{i\omega} - 1\right ) = e^{\frac{i\omega}{2}} \Leftrightarrow
\\
   & H(\omega) = \frac{1}{e^{\frac{i\omega}{2}} - e^{-\frac{i\omega}{2}}} = \frac{1}{2i\sin\frac{\omega}{2}}.
\end{align*}
$$

Воспользуемся связью передаточной функции для круговой частоты и циклической, получим: 
$$
\tilde H(f) = H(2\pi f) = \frac{1}{2i\sin (\pi f)}.
$$
Точное значение посчитаем как 
$$
  y(t) = \int s(t)dt = \int e^{i\omega t}dt = \frac{e^{i\omega t}}{i\omega},\ (2)
$$
вычисленное значение, соответственно:
$$
   y_n = \frac{e^{i\omega n}}{2i\sin\frac{\omega}{2}},
$$

тогда отношение вычисленного значения к точному при t = n равно:

$$
   \gamma = \frac{i\omega}{2i\sin\frac{\omega}{2}} = \frac{\frac{\omega}{2}}{\sin\frac{\omega}{2}} = \frac{\pi f}{\sin(\pi f)}
$$

#### Квадратурная формула трапеций 
Выведем формулу передаточной функции рекурсивного фильтра, соответствующей квадратурной формуле прямоугольников $ y_{n+1} = y_n + \frac{1}{2}(s_n + s_{n+1}),\ (3)$ где $y_0 = 0.$ Отсчет входного сигнала как $s_n=e^{i\omega n},$ выход фильтра $y_n=H(\omega)e^{i\omega n}.$ Тогда, подставляя эти выражения в уравнение (3), получим:
$$
\begin{align*}
   & H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega n} + \frac{e^{i\omega n} + e^{i\omega (n+1)}}{2} \Leftrightarrow
   H(\omega)e^{i\omega n} e^{i\omega} = H(\omega)e^{i\omega n} + e^{i\omega n}\frac{1 + e^{i\omega}}{2} \Leftrightarrow
\\
   & H(\omega) (e^{i\omega} - 1)= \frac{1 + e^{i\omega}}{2} \Leftrightarrow
   H(\omega) = \frac{e^{i\omega + 1}}{2(e^{i\omega}-1)} = \frac{\cos\frac{\omega}{2}}{2i\sin\frac{\omega}{2}} \Rightarrow
\\
   & \tilde{H}(f) = \frac{\cos(\pi f)}{2i\sin{\pi f}}
\end{align*}
$$
Пользуясь выражением (2), найдем отношение вычисленного значения к точному при t = n:

$$
   \gamma = \frac{\left(\frac{e^{i\omega n}\cos\frac{\omega}{2}}{2i\sin\frac{\omega}{2}}\right )}{\frac{e^{i\omega n}}{i\omega}} = \frac{i\omega\cos\frac{\omega}{2}}{2i\sin\frac{\omega}{2}} = \frac{\omega}{2}\cdot \frac{\cos\frac{\omega}{2}}{\sin\frac{\omega}{2}} = \frac\omega 2 \ctg\frac\omega 2 = \pi f \ctg \pi f
$$

#### Формула Симпсона
Выведем формулу передаточной функции рекурсивного фильтра, соответствующей формуле Симпсона $ y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_n + s_{n+1}),\ (4)$ где $y_0 = 0.$ Отсчет входного сигнала как $s_n=e^{i\omega n},$ выход фильтра $y_n=H(\omega)e^{i\omega n}.$ Тогда, подставляя эти выражения в уравнение (4), получим:
$$
\begin{align*}
   & H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega (n-1)} + \frac{e^{i\omega (n - 1)} + 4e^{i\omega n} + e^{i\omega (n+1)}}{3} \Leftrightarrow
   H(\omega) e^{i\omega} = H(\omega)e^{-i\omega} + \frac{e^{-i\omega} + 4 + e^{i\omega}}{3} \Leftrightarrow
\\
   & H(\omega) (e^{i\omega} - e^{-i\omega})= \frac{e^{-i\omega} + 4 + e^{i\omega}}{3} \Leftrightarrow
   H(\omega) = \frac{e^{-i\omega} + 4 + e^{i\omega}}{3(e^{i\omega} - e^{-i\omega})} = \frac{2 + \cos\omega}{3i\sin\omega} \Rightarrow
\\
   & \tilde{H}(f) = \frac{2 + \cos (2\pi f)}{3i\sin (2\pi f)}.
\end{align*}
$$
Пользуясь выражением (2), найдем отношение вычисленного значения к точному при t = n:

$$
   \gamma = \frac{e^{i\omega n}\left(\frac{2 + \cos\omega}{3i\sin\omega}\right )}{\frac{e^{i\omega n}}{i\omega}} = \frac{(2+\cos\omega)\omega}{3\sin\omega} = \frac{2\pi f (2 + \cos (2\pi f))}{3\sin(2\pi f)}
$$

#### Построение графиков
Построим графики передаточных функций и графики отношения вычисляемого в результате фильтрации значения к истинному. При построении графиков передаточной функции, строится только мнимая часть, т.к. действительная равна 0. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H_rect = lambda f: 1 / (2j * np.sin(np.pi * f))
H_trap = lambda f: 1 / (2j * np.tan(np.pi *f))
H_simp = lambda f: (2 + np.cos(2 * np.pi * f)) / (3j * np.sin(2 * np.pi * f)) 

g_rect = lambda f: np.pi * f / (np.sin(np.pi * f))
g_trap = lambda f: np.pi * f / (np.tan(np.pi * f)) 
g_simp = lambda f: (2 + np.cos(2 * np.pi * f)) / 3 * 2 * np.pi * f / (np.sin(2 * np.pi * f))

F = np.linspace(-0.5, 0.5, 2**8)
get_f_x = lambda func: [x.imag if x is not None else None for x in [func(f) if np.abs(f) > 0.05 else None for f in F]]
get_g_x = lambda func: [x for x in [func(f) for f in F]]
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plt.plot(F, get_f_x(H_rect), label='Формула прямоугольников', color='blue')
plt.plot(F, get_f_x(H_trap), label='Формула трапеций', color='green')
plt.plot(F, get_f_x(H_simp), label='Формула Симпсона', color='red')

plt.xlabel('f')
plt.ylabel('$H(f)$')
plt.title("Передаточные функции фильтров")
plt.legend()
plt.xticks([round(x, 2) for x in np.linspace(-0.5, 0.5, 10)])
plt.ylim((-3, 3))
plt.grid()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plt.plot(F, get_g_x(g_rect), label='Формула прямоугольников', color='blue')
plt.plot(F, get_g_x(g_trap), label='Формула трапеций', color='green')
plt.plot(F, get_g_x(g_simp), label='Формула Симпсона', color='red')

plt.xlabel('f')
plt.ylabel('$H(f)$')
plt.title("Отношения вычисленных значений к реальным")
plt.legend()
plt.xticks([round(x, 2) for x in np.linspace(-0.5, 0.5, 10)])
plt.ylim((0, 2.5))
plt.grid()
plt.show()
```

#### Интерпретация частотных свойств полученных передаточных функций
- Формула прямоугольников: усиливаются низкие частоты, высокие подавляются
- Формула трапеций: усиливаются низкие частоты, высокие немного подавляются
- Формула Симпсона: усиливаются низкие частоты, высокие частоты усиливаются сильнее

Все фильтры усиливают низкие частоты. 

### Задание 2
Вывести формулу передаточной функции рекурсивного фильтра для интегрирования «по правилу 3/8». Построить график передаточной функции и график отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства передаточной функции.

#### Рекурсивный фильтр для интегрирования «по правилу 3/8»
Выведем формулу передаточной функции рекурсивного фильтра, соответствующей формуле интегрирования «по правилу 3/8» $ y_{n+2} = y_{n-1} + \frac{1}{8}(s_{n+2} + 3s_{n+1} + 3s_n + s_{n-1}),\ (5)$ где $y_0 = 0.$ Отсчет входного сигнала как $s_n=e^{i\omega n},$ выход фильтра $y_n=H(\omega)e^{i\omega n}.$ Тогда, подставляя эти выражения в уравнение (5), получим:
$$
\begin{align*}
   & H(\omega)e^{i\omega (n+2)} = H(\omega)e^{i\omega (n-1)} + \frac{1}{8}\left (e^{i\omega(n+2)} + 3e^{i\omega (n+1)} + 3e^{i\omega n} + e^{i\omega (n-1)} \right) \Leftrightarrow
   H(\omega)e^{2i\omega} = H(\omega)e^{-i\omega} + \frac{1}{8}\left (e^{2i\omega} + 3e^{i\omega} + 3 + e^{-i\omega} \right) \Leftrightarrow
\\
   & H(\omega) (e^{2i\omega} - e^{-i\omega})= \frac{1}{8}\left (e^{2i\omega} + 3e^{i\omega} + 3 + e^{-i\omega} \right) \Leftrightarrow
   H(\omega) = \frac{e^{2i\omega} + 3e^{i\omega} + 3 + e^{-i\omega}}{8(e^{2i\omega} - e^{-i\omega})} = 
   \frac{2\left (\frac{e^{\frac{3i\omega}{2}} + e^{\frac{-3i\omega}{2}}}{2} \right) + 2\left (\frac{e^{\frac{i\omega}{2}} + e^{\frac{-i\omega}{2}}}{2} \right)}{8\cdot 2\cdot\frac{e^{\frac{3i\omega}{2}} - e^{\frac{-3i\omega}{2}}}{2}} =
   \frac{\cos\frac{3\omega}{2} + 3\cos\frac\omega 2}{8i\sin\frac{3\omega}2}
   \Rightarrow
\\
   & \tilde{H}(f) = \frac{\cos(3\pi f) + 3\cos(\pi f)}{8i\sin (3\pi f)}
\end{align*}
$$
Пользуясь выражением (2), найдем отношение вычисленного значения к точному при t = n:

$$
   \gamma = \frac{e^{i\omega n\cdot\frac{\cos\frac{3\omega}{2} + 3\cos\frac\omega 2}{8i\sin\frac{3\omega}2}}}{{\frac{e^{i\omega n}}{i\omega}}} = \omega\frac{\cos\frac{3\omega}{2} + 3\cos\frac\omega 2}{8\sin\frac{3\omega}2} = \pi f \frac{\cos(3\pi f) + 3\cos(\pi f)}{4\sin (3\pi f)}
$$

#### Построение графиков
Построим график передаточной функций и график отношения вычисляемого в результате фильтрации значения к истинному. При построении графика передаточной функции, строится только мнимая часть, т.к. действительная равна 0.

```python
H = lambda f: (np.cos(3 * np.pi * f) + 3 * np.cos(np.pi * f))/(8j * np.sin(3 * np.pi * f))
g = lambda f: (np.pi * f)/4 * (np.cos(3 * np.pi * f) + 3 * np.cos(np.pi * f))/(4 * np.sin(3 * np.pi * f))

F = np.linspace(-0.5, 0.5, 2**16)

in_rad = lambda f, x0: np.abs(f-x0) < 0.0005

get_f_x = lambda func: [func(f).imag if f is not None else None for f in [None if in_rad(3*f, 1) or in_rad(3*f, -1) or in_rad(f, 0) else f for f in F]]
get_g_x = lambda func: [func(f) if f is not None else None for f in [None if in_rad(3*f, 1) or in_rad(3*f, -1) else f for f in F]]
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plt.plot(F, get_f_x(H), label='Формула интегрирования правилом 3/8', color='blue')

plt.xlabel('f')
plt.ylabel('$H(f)$')
plt.title("Передаточная функция фильтра")
plt.xticks([round(x, 2) for x in np.linspace(-0.5, 0.5, 10)])
plt.ylim((-3, 3))
plt.grid()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plt.plot(F, get_g_x(g), label='Формула прямоугольников', color='blue')

plt.xlabel('f')
plt.ylabel('$H(f)$')
plt.title("Отношения вычисленного значения к реальному")
plt.xticks([round(x, 2) for x in np.linspace(-0.5, 0.5, 10)])
plt.ylim((-4, 4))
plt.grid()
plt.show()
```

Из графиков видно, что подавляются все частоты, кроме $f = \pm\frac{1}{3}$ и $ f = 0,$ эти частоты усиливаются
