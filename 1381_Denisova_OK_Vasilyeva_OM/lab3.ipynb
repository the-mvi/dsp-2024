### Дискретный нерекурсивный фильтр
Для заданного дискретного сигнала $x(n), n \in [0, N - 1]$ последовательность, описывающая выходной сигнал нерекурсивного фильтра: $ y_{n} = \sum_{k=0}^M{[c_{k}x_{n-k}]}$, где M - размер окна, в котором производится сглаживание

### Передаточная функция
Для дискретного сигнала $x_{n}$ передаточная функция рекурсивного фильтра вычисляется как:
$H(w) = \frac{y_{n}}{x_{n}}$

Применяя к данному равенству формулу последовательности выходного сигнала дискретного нерекурсивного фильтра получим:
$ H(w) = \frac{\sum_{k=0}^M{[c_{k}e^{iw(n-k)}]}}{e^{iwn}} = \sum_{k=0}^M{[c_{k}e^{iw(n-k)}] * e^{-iwn}} $
$ \newline$

### Связь передаточной фунции для круговой частоты и циклической

$ H(w) = H(2*\pi * f) =  \widetilde{H}(f)  $

### Задание 1
Смоделировать дискретный прямоугольный импульс: 
Вывести формулы для передаточной функции нерекурсивного фильтра, соответствующего сглаживанию прямой линией по 3, 5, 7 и 9 точкам. Построить графики $\widetilde{H}(f)$. Проинтерпретировать частотные свойства передаточных функций для различного количества точек.

Входной сигнал: s(t).
Выходной сигнал: yt=A+Bt.
Если осуществлять приближение по МНК прямой линией по m точкам, то необходимо минимизировать следующее выражение для нечетных чисел: <br>
$F(A,B) = \sum_{k=-m}^{m}(s_k - y_k)^2= \sum_{k=-m}^{m}(s_k - A - Bk)^2 \Rightarrow min$ <br>
Переменными в этой задаче являются коэффициенты A и B прямой линии. Продифференцируем и приравняем к нулю, чтобы найти минимум. Посчитаем частные производные по A и B, получим: <br>
$ \begin{cases}
\frac{\delta F(A, B)}{\delta A} = 0 \\
\frac{\delta F(A, B)}{\delta B} = 0 
\end{cases}$ $\iff$ 

$\begin{cases}
\ - 2\sum_{k=-m}^{m}(s_k - A - Bk) = 0 \\
\ -2\sum_{k=-m}^{m}(k s_k - kA - Bk^2) = 0 
\end{cases}$ $\iff$

$\begin{cases}
\ -\sum_{k=-m}^{m}s_k + \sum_{k=-m}^{m}A + \sum_{k=-m}^{m}Bk = 0 \\
\ -\sum_{k=-m}^{m}k s_k  + \sum_{k=-m}^{m}kA + \sum_{k=-m}^{m}Bk^2= 0 
\end{cases}$

Тогда система нормальных уравнений:
$ \begin{cases}
\ A = \frac{1}{2m + 1}\sum_{k=-m}^{m}s_k \\
\ B = \frac{\sum_{k=-m}^{m}s_k}{\sum_{k=-m}^{m}k^2}
\end{cases}$

В итоге получаем:
$y_0 = A = \frac{1}{2m + 1}\sum_{k=-m}^{m}s_k = \frac{1}{2m + 1}\sum_{k=-m}^{m}(s_{-m} + s_{-m+1}+ ... + s_{m-1} + s_m)$

В общем случае:

$y_n = \frac{1}{2m+1}\sum_{k=-m}^{m}s_k = \frac{1}{2m + 1}\sum_{k=-m}^{m}(s_{-m+n} + s_{-m+1+n}+ ... + s_{m-1+n} + s_{m+n})$

$s_n= e^{i\omega n}$

$y_n = \frac{1}{2m + 1}\sum_{k=-m}^{m}(e^{-mi\omega}+e^{(-m+1)\omega}+ ... +e^{(m-1)\omega}+e^{mi\omega}) = H(\omega)$

$H(\omega)= \frac{\sin{(\frac{(2m+1)\omega}{2})}}{(2m+1)\sin{(\frac{\omega}{2})}}$

$H(\omega) = H(2\pi f) = \widetilde{H}(f)$

Формула для передаточной функции нерекурсивного фильтра,
соответствующего сглаживанию прямой линией по 3 точкам:
$H_3(f) = \frac{\sin{(3\pi f)}}{3 \sin{(\pi f)}}$

Формула для передаточной функции нерекурсивного фильтра,
соответствующего сглаживанию прямой линией по 5 точкам:
$H_5(f) = \frac{\sin{(5\pi f)}}{5 \sin{(\pi f)}}$

Формула для передаточной функции нерекурсивного фильтра,
соответствующего сглаживанию прямой линией по 7 точкам:
$H_7(f) = \frac{\sin{(7\pi f)}}{7 \sin{(\pi f)}}$

Формула для передаточной функции нерекурсивного фильтра,
соответствующего сглаживанию прямой линией по 9 точкам:
$H_9(f) = \frac{\sin{(9\pi f)}}{9 \sin{(\pi f)}}$

Для построения графиков возьмем интервал $ f \in [0, 0.5] $



```python
import numpy as np
import math
import matplotlib.pyplot as plt

h_3 = lambda x: np.sin(3*math.pi*x)/(3*np.sin(x*math.pi))
h_5 = lambda x: np.sin(5*math.pi*x)/(5*np.sin(x*math.pi))
h_7 = lambda x: np.sin(7*math.pi*x)/(7*np.sin(x*math.pi))
h_9 = lambda x: np.sin(9*math.pi*x)/(9*np.sin(x*math.pi))

t = np.linspace(0.001, 1.0, 64)
plt.plot(t, h_3(t), label=r'$H_3$')
plt.plot(t, h_5(t), label=r'$H_5$')
plt.plot(t, h_7(t), label=r'$H_7$')
plt.plot(t, h_9(t), label=r'$H_9$')
plt.xlabel('t')
plt.ylabel(r'$H(t)$')
plt.legend()
plt.title('Сглаживание прямой по методу наименьших квадратов.' )
plt.show()
```

Можно заметить, что каждый период каждой передаточной функции содержит ровно столько экстремумов, сколько было использовано точек для её построения. Соответственно, большее количество колебаний наблюдается при большем количестве экстремумов. Кроме того, увеличение количества точек для построения функции приводит к повышению порядка гладкости.
