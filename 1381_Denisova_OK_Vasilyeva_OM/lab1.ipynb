## Основные теоретические положения
Цифровые сигналы представляют собой дискретные формы сигналов, которые изменяются через определенные промежутки времени и могут быть представлены последовательностью значений. 

Аналоговые сигналы $s(t)$, в отличие от цифровых, являются непрерывными и могут принимать любое значение в пределах заданного диапазона. 

Дискретные сигналы $s(nT)$ формируются путем выборки аналогового сигнала в определенные моменты времени. 

Частота дискретизации указывает на то, с какой частотой аналоговый сигнал преобразуется в дискретный.

Единичный цифровой импульс $ \delta_d(k) $:

$
\delta_d(k) = 
\begin{cases} 
1, & k = 0 \\ 
0, & k \neq 0 
\end{cases}, \; k \in \mathbb{Z}
$

Дельта-функция $ \delta(t) $:

$
\delta(t) = 
\begin{cases} 
\infty, & t = 0 \\ 
0, & t \neq 0 
\end{cases}, \; t \in \mathbb{R}
$
с условием:

$ \int_{-\infty}^{\infty} \delta(t) \, dt = 1 $


Дискретный скачок $ \sigma_d(k) $:

$
\sigma_d(k) = 
\begin{cases} 
1, & k \geq 0 \\ 
0, & k < 0 
\end{cases}, \; k \in \mathbb{Z}
$

Аналоговый скачок $ \sigma(t) $:

$
\sigma(t) = 
\begin{cases} 
1, & t > 0 \\ 
0.5 \text{ или не определена}, & t = 0 \\ 
0, & t < 0 
\end{cases}, \;  t \in \mathbb{R}
$

Частота дискретизации $ f_d $ определяется как обратная величина периода дискретизации $ T $:

$
f_d = \frac{1}{T}
$

Частота дискретизации позволяет понять, с какой частотой аналоговый сигнал преобразуется в дискретный.

Дискретная экспонента задается формулой:

$
s_1(k) = 
\begin{cases} 
a^k, & k \geq 0 \\ 
0, & k < 0 
\end{cases}, \;  k \in \mathbb{Z}
$

Аналоговая экспонента задается:

$
s(t) = 
\begin{cases} 
a^t, & t \geq 0 \\ 
0, & t < 0 
\end{cases}, \;  t \in \mathbb{R} 
$

Дискретный комплексный гармонический сигнал $ s_2(k) $ задается следующим образом, где $ k \in \mathbb{Z}$:

$
s_2(k) = C \exp(j \hat{\omega_0} k)
$

Этот сигнал можно записать в виде комбинации двух вещественных последовательностей с использованием формулы Эйлера:

$
s_2(k) = C \cdot (\cos(\hat{\omega_0} k) + j \sin(\hat{\omega_0} k))
$

## Выполнение работы

```python
# Основные переменные

import numpy as np
import math
import matplotlib.pyplot as plt
Nb = 2 # Номер бригады
N = 30 + Nb % 5 # Длина последовательности
T = 0.0005*(1 + Nb % 3) # Период дискретизации
a = (-1) ** Nb * (0.8 + 0.005 * (Nb % 20))  # Основание экспоненты
C = 1 + Nb % 5  # Амплитуда гармонического сигнала
w0 = math.pi / (6 + Nb % 5)  # Частота гармонического сигнала
m = 5 + Nb % 5  # Задержка

print("Nb:", Nb)
print("N:", N)
print("T:", T)
print("a:", a)
print("C:", C)
print("w0:", w0)
print("m:", m)

```

### Задание 1

Смоделировать единичный цифровой импульс $ \delta_d(k) $ с выводом графиков на интервале дискретного времени $ nT \in [0, (N-1)T] $ и дискретного нормированного времени $ n \in [0, N-1] $. Пояснить: <br>
 - Взаимосвязь между дискретным и дискретным нормированным временем <br>
 - Различие между единичным цифровым импульсом и дельта-функцией

```python
x, y = [], []

# Единичный цифровой импульс

for i in np.arange(0, (N-1)*T, 0.003):
    x.append(i)
    if i == 0:
        y.append(1)
    else:
        y.append(0)
plt.title('Единичный цифровой импульс δ_d(k) (nT)')
plt.ylabel('Амплитуда')
plt.vlines(x, ymin=0, ymax=y, color='firebrick', alpha=0.7, linewidth=2)
plt.scatter(x, y, s=75, color='firebrick', alpha=0.7)
plt.show()

x, y = [], []

for i in np.arange(0, N - 1, 2):
    x.append(i)
    if i == 0:
        y.append(1)
    else:
        y.append(0)
plt.title('Единичный цифровой импульс δ_d(k) (n)')
plt.ylabel('Амплитуда')
plt.vlines(x, ymin=0, ymax=y, color='firebrick', alpha=0.7, linewidth=2)
plt.scatter(x, y, s=75, color='firebrick', alpha=0.7)
plt.show()
```

a. Взаимосвзяь между дискретным временем и дискретным нормированным временем <br>

nT -  дискретное время (время n-ого отсчета), а n — дискретное нормированное время (T = 1).

b. Различия между единичным цифровым импульсом и дельта-функцией <br>

Различие между цифровым единичным импульсом $\delta_d$ и $ \delta$ -функцией состоит в том, что цифровой единичный импульс является физически реализуемым сигналом, тогда как аналоговый единичный импульс рассматривается только как обобщённая функция.

В единичном цифровом импульсе амплитуда равна единицы, тогда как у Дельта-функции - бесконечности.

Единичный цифровой импульс $ \delta_d(k) $

1, &  k = 0 \\
0, &  k \neq 0
\end{cases}$,  $\; k \in \mathbb{Z}$

### Задание 2
Смоделировать дискретный единичный скачок $ \sigma_d(k) $ с выводом графиков на интервале дискретного времени $ nT \in [0, (N-1)T] $ и дискретного нормированного времени $ n \in [0, N-1] $. Пояснить:
- Соответствие между дискретным и аналоговым единичными скачками.
- Чему равна частота дискретизации дискретного единичного скачка.


```python
x, y = [], []

# Дискретный единичный скачок

for i in np.arange(0, (N-1)*T, 0.003):
    x.append(i)
    if i < 0:
        y.append(0)
    else:
        y.append(1)
plt.title('Дискретный единичный скачок σ_d(k) (nT)' )
plt.ylabel('Амплитуда')
plt.vlines(x, ymin=0, ymax=y, color='firebrick', alpha=0.7, linewidth=2)
plt.scatter(x, y, s=75, color='firebrick', alpha=0.7)
plt.show()

x, y = [], []

for i in np.arange(0, N - 1, 2):
    x.append(i)
    if i < 0:
        y.append(0)
    else:
        y.append(1)

plt.title('Дискретный единичный скачок σ_d(k) (n)')
plt.ylabel('Амплитуда')
plt.vlines(x, ymin=0, ymax=y, color='firebrick', alpha=0.7, linewidth=2)
plt.scatter(x, y, s=75, color='firebrick', alpha=0.7)
plt.show()
```

a. Соответствие между дискретным и аналоговым единичными скачками

Дискретный единичный скачок - это дискретный аналог функции единичного скачка (функции Хэвисайда).

$\sigma_d(k) = \begin{cases}
1, &  k \geq 0 \\
0, &  k < 0
\end{cases}$,   $\; k \in \mathbb{Z}$

Аналоговый единичный скачок

$\sigma(t) = \begin{cases}
1, &  t > 0 \\
0.5  \text{ или не определена}, &  t = 0 \\ 
0, &  t < 0
\end{cases}$ ,  $\; t \in \mathbb{R}$

b. Частота дискретизации дискретного единичного скачка

Частоту дискретизации дискретного единичного скачка можно найти 
$ f_d = \frac{1}{T} = \frac{1}{0.0015} ≈ 666.66667 \; \text{Гц}$

где $ T $ — период дискретизации.

### Задание 3
Смоделировать дискретную экспоненциальную функцию $ s_1(k) $ с выводом графиков на интервале дискретного времени $ nT \in [0, (N-1)T] $ и дискретного нормированного времени $ n \in [0, N-1] $. Пояснить соответствие между дискретной и аналоговой экспонентами.

```python
nT = np.arange(0, N) * T # Дискретное время
n = np.arange(0, N) # Дискретное нормированное время

s1 = a ** n

plt.figure(figsize=(7, 4))
plt.stem(nT, s1, markerfmt="o", linefmt='firebrick')
plt.title('Дискретная экспоненциальная функция s_1(k) (nT)')
plt.ylabel('Амплитуда')
plt.show()

x, y = [], []

for i in np.arange(0, N - 1, 1):
    x.append(i)
    if i < 0:
        y.append(0)
    else:
        y.append(a ** i)
plt.title('Дискретная экспоненциальная функция s_1(k) (n)')
plt.ylabel('Амплитуда')
plt.vlines(x, ymin=0, ymax=y, color='firebrick', alpha=0.7, linewidth=2)
plt.scatter(x, y, s=75, color='firebrick', alpha=0.7)
plt.show()
```

Соответствие между дискретной и аналоговой экспонентой <br>

Дискретная экспонента

$s_1(k) = \begin{cases} 
a^k, & k \geq 0 \\ 
0, &  k < 0 
\end{cases}$ ,  $\; k \in \mathbb{Z}$

Аналоговая экспонента

$s(t) = \begin{cases} 
a^t, &  t \geq 0 \\ 
0, &  t < 0 
\end{cases}$ ,  $\; t \in \mathbb{R}$

Точки дискретной экспоненты находятся в местах, где для аналоговой
экспоненты $a^k$
, k – целые.

Дискретная экспонента (экспоненциальная последовательность)
образуется в результате дискретизации экспоненты.

### Задание 4
Смоделировать дискретный комплексный гармонический сигнал $s_2(k) = C \exp(j \hat{\omega_0} k)$ с выводом графиков вещественной и мнимой частей на интервале времени $ n \in [0, N-1] $ Записать данный сигнал в виде комбинации двух вещественных последовательностей.

```python
s2 = C * np.exp(1j * w0 * n) # Дискретный комплексный гармонический сигнал
real = np.real(s2) # Вещественная часть
imag = np.imag(s2) # Мнимая часть

plt.figure(figsize=(10,3))
plt.stem(n, real, linefmt='firebrick')
plt.title('Вещественная часть комплексного гармонического сигнала')
plt.xlabel('n')
plt.ylabel('Амплитуда')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 3))
plt.stem(n, imag, linefmt='firebrick')
plt.title('Мнимая часть комплексного гармонического сигнала')
plt.xlabel('n')
plt.ylabel('Амплитуда')
plt.grid(True)

plt.show()
```

Данный сигнал можно записать в виде двух вещественных последовательностей: <br>

$ Re(x(k)) = C cos(w0k)$ <br>
$ Im(x(k)) = C sin(w0k)$

$s_2Re(k) = 3cos(w0k)$ <br>
$s_2Im(k) = j3sin(w0k)$ <br>

$s_2Re(k) = 3cos(\frac{\pi k}{8})$ <br>
$s_2Im(k) = j3sin(\frac{\pi k}{8})$
