## Основные теоретические положения
#### Дискретный единичный скачок



 $ σ_{d}(k) = \begin{equation}
\left\{
   \begin{align*} 
  1, k \geqslant 0 \\ \nonumber
  0, k < 0\\ \nonumber
  \end{align*} \right.
\end{equation} $

#### Дискретный прямоугольный импульс
 $ s_{1}(k) = \begin{equation}
\left\{
   \begin{aligned} 
  U, n_{0} \leqslant k  \leqslant  n_{0} + n_{imp} - 1\\ \nonumber
  0,   else\\ \nonumber
  \end{aligned} \right.
\end{equation} $


#### Линейная комбинация дискретных гармонических сигналов $ s_{2}(k) $

$ s_{2}(k) = a_{1}x_{1}(k) + a_{2}x_{2}(k) + a_{3}x_{3}(k)$, где $ x_{i}(k) = B_{i}sin(ŵ_{i}k), i = 1,2,3... $ 

#### Дискретный гармонический сигнал $ s_{3}(k) $

$ s_{3}(k) = C * cos(ŵ_{0}k) $

## Выполнение работы

```python
# Основные переменные

import numpy as np
import math
import matplotlib.pyplot as plt
Nb = 2 # Номер бригады
N = 30 + Nb % 5 # Длина последовательности
a = (-1) ** Nb * (0.8 + 0.005 * (Nb % 20))  # Основание экспоненты
C = 1 + Nb % 5  # Амплитуда гармонического сигнала
w0 = math.pi / (6 + Nb % 5)  # Частота гармонического сигнала
U = Nb % 20 +1 # Амплитуда импульса
n0 = 3 + Nb % 5 # Начальный момент импульса
n_imp = 5 + Nb % 5 # Длина импульса
# Амплитуды гармонических сигналов
B1 = 1.5 + Nb % 5 
B2 = 5.7 - Nb % 5 
B3 = 2.2 + Nb % 5 
# Частоты гармонических сигналов
w1 = math.pi / (4 + Nb % 5) 
w2 = math.pi / (8 + Nb % 5) 
w3 = math.pi / (16 + Nb % 5)
# Коэффициенты линейной комбинации гармонических сигналов
a1 = 1.5 - Nb % 5 
a2 = 0.7 + Nb % 5 
a3 = 1.4 + Nb % 5 

print("Nb:", Nb)
print("N:", N)
print("a:", a)
print("C:", C)
print("w0:", w0)
print("U:", U)
print("n0:", n0)
print("n_imp:", n_imp)
print("B1:", B1, " B2:", B2, " B3:", B3)
print("w1:", w1, " w2:", w2, " w3:", w3)
print("a1:", a1, " a2:", a2, " a3:", a3)

```

### Задание 1
Смоделировать дискретный прямоугольный импульс: 
$s_1(k) = \begin{cases} 
U, & n_0 \leq k \leq n_0 + n_{imp} +1\\ 
0, &  иначе 
\end{cases}$ <br>
на основе дискретного единичного скачка $\sigma_d(k)$  из л/р №1 с выводом графика на интервале времени $ n \in [0, N-1] $ Пояснить как выполняется моделирование импульса.

```python
n = np.arange(0, N) # Дискретное нормированное время

def rect (_x, m = n0):
    y = np.zeros(_x.shape)
    y[(m <= _x) & (_x <= m + n_imp - 1)] = U
    return y

plt.title('Дискретный прямоугольный импульс s_1(k)' )
plt.stem(n, rect(n), markerfmt="o", linefmt='firebrick')
plt.ylabel(r'$s_{1}$(n)')
plt.xlabel('n')
plt.show()
```

Моделирование прямоугольного импульса происходит следующим образом:
1. Генерируется массив из нулей с количеством элементов соответствующем интервалу времени.
2. Каждому элементу, удовлетворяющему условию, присваивается U.


### Задание 2
Смоделировать линейную комбинацию дискретных гармонических сигналов $s_2(k)$: <br>
$ s_2(k) = a_1x_1(k) + a_2x_2(k) + a_3x_3(k) $ , где
$ x_i(k) = B_i\sin(\hat{\omega_i} k) , i = 1, 2, 3$
с выводом графиков последовательностей $x_i(k)$ и $s_2(k)$ на интервале времени $ n \in [0, 5N-1] $. Вычислить среднее значение (идентификатор mean_s2), энергию (идентификатор E) и среднюю мощность последовательности $s_2(k) $
(идентификатор P). Пояснить:
- операции при моделировании линейной комбинации сигналов $ s_2(k)$;
- как определяют указанные характеристики.



```python
x1 = lambda k: B1 * np.sin(w1 * k)
x2 = lambda k: B2 * np.sin(w2 * k)
x3 = lambda k: B3 * np.sin(w3 * k)

n2 = np.arange(0, 5*N) 

def s4(_x):
    return a1 * x1(_x) + a2 * x2(_x) + a3 * x3(_x)

# s_2(k)
ans = s4(n2)
plt.title('График последовательности s_2(k)' )
plt.stem(n2, ans, markerfmt="o", linefmt='firebrick')
plt.ylabel(r'$s_{2}$(n)')
plt.xlabel('n')
plt.show()

# x_1(k)
plt.title('График последовательности x_1(k)' )
plt.stem(n2, x1(n2), markerfmt="o", linefmt='firebrick')
plt.ylabel(r'$x_{1}$(n)')
plt.xlabel('n')
plt.show()

# x_2(k)
plt.title('График последовательности x_2(k)' )
plt.stem(n2, x2(n2), markerfmt="o", linefmt='firebrick')
plt.ylabel(r'$x_{2}$(n)')
plt.xlabel('n')
plt.show()

# x_3(k)
plt.title('График последовательности x_3(k)' )
plt.stem(n2, x3(n2), markerfmt="o", linefmt='firebrick')
plt.ylabel(r'$x_{3}$(n)')
plt.xlabel('n')
plt.show()

#Вычислим указанные в задании характеристики
mean_s2 = ans.mean() # среднее занчение
E = np.power(ans, 2).sum() # Энергия
P = np.power(ans, 2).mean() # Средняя мощность последовательности
print('mean_s2 = {}, E = {}, P = {}'.format(mean_s2, E, P))
```

a. Операции при моделировании линейной комбинации сигналов:
1. Вычисление дискретного нормированного времени
2. Вычисление матрицы дискретных гармоник
3. Линейная комбинация дискретных гармоник

b. Определение характеристик:
- Cреднее значение: mean_s2 = $ \frac{1}{N}\sum_{i=0}^{N-1}x_i$
- Энергия: E = $\sum x^2$
- Мощность: P = $\frac{\sum x^2}{N}$


### Задание 3
Вывести график дискретного сигнала $s_3(k)$ представляющего собой дискретный гармонический сигнал s(k) <br>
s(k) = $C \cos(\hat{\omega_0} k)$
с экспоненциальной огибающей $ |a|^k $, на интервале времени $ n \in [0, N-1] $. Привести аналитическую формулу дискретного сигнала s3(k) и пояснить операции при его моделировании.

```python
n3 = np.arange(0, N)

def s3(_x):
    return C * math.cos(w0 * _x)

def ak(_x):
    return abs(a) ** _x

ans = np.vectorize(s3)
ans2 = np.vectorize(ak)

plt.title('График дискретного гармонического сигнала $ s_{3}(k) $  и экспоненциальной огибающей $ |a|^{k} $' )
plt.stem(n3, ans(n3) * ans2(n3), markerfmt="o", linefmt='firebrick', label="|a|^k * s_3(k)")
plt.stem(n3, ans2(n3), markerfmt="o", label="s_3(k)")
plt.ylabel(r'$s_{3}$(k)')
plt.xlabel('n')
plt.legend(loc='upper right')
plt.show()
```

Для того, чтобы сигнал $ |a|^k $ огибал сигнал $ s_{3}(k) $, необходимо сигнал  $ s_{3}(k) $ домножить на  $ |a|^k $, таким образом получаем дискретный гармонический сигнал с экпоненциальной огибающей:
$ s_{3}(k) = |a|^{k} * C * cos(w_{0} * k)  $

### Задание 4
Вывести график пяти периодов периодической последовательности $s_4(k)$ дискретных прямоугольных импульсов амплитуды U и длительности $n_{imp}$ с периодом, вдвое большим длительности импульса. Пояснить операции при моделировании периодической последовательности.

```python
n4 = np.arange(0, 3*N) 
y = np.zeros(shape=n4.shape)

for i in range(5):
    y += rect(n4, 2 * n_imp * i)

plt.title('График пять периодов периодической последовательности s_4(k)')
plt.stem(n4, y, markerfmt="o", linefmt='firebrick')
plt.ylabel(r'$s_{4}$(n)')
plt.xlabel('n')
plt.show()
```

Операции при моделировании периодической последовательности:

2. Складываем их.
