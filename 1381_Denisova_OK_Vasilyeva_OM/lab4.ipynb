### Задание 1
Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона. Построить графики передаточных функций и графики отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства полученных передаточных функций.

Формула прямоугольников:<br>
$y_{n+1} = y_n + s_{n + \frac{1}{2}}, y_0 = 0$ <br>

Пусть $s_n = e^{i \omega n}$ и $y_n = H(\omega)e^{i \omega n} $, тогда<br>
$ \begin{cases}
y_{n+1} = H(\omega)e^{i \omega n} + e^{i \omega (n + \frac{1}{2})} \\
y_{n+1} = H(\omega)e^{i \omega (n+1)}
\end{cases}$ <br>

$H(\omega)(e^{i \omega n}e^{i \omega}) = H(\omega)e^{i \omega n} + e^{i \omega n}e^{\frac{1}{2}i \omega}$ <br>

$H(\omega)(e^{i \omega n}e^{i \omega} - e^{i \omega n}) = e^{i \omega n}e^{\frac{1}{2}i \omega}$<br>

$H(\omega)(e^{i \omega} - 1) = e^{\frac{1}{2}i \omega}$<br>

$H(\omega) = \frac{1}{e^{\frac{i \omega}{2}} - e^{-\frac{i \omega}{2}}} = \frac{1}{2i \sin{(\pi f)}}$<br>

$\widetilde{H}(f) = \frac{1}{2i \sin{(\pi f)}}$<br>

Точное значение интеграла $e^{i \omega t}$ равно $\frac{e^{i \omega t}}{i \omega}$:<br>
$\gamma = \frac{Вычисленное}{Точное} = \frac{i\omega}{2i\sin{(\frac{\omega}{2})}} = \frac{\frac{\omega}{2}}{\sin{(\frac{\omega}{2})}}$ <br> 

Формула трапеции:<br>
$y_{n+1} = y_n + \frac{1}{2}(s_n + s_{n+1}), y_0 = 0 $ <br>

Пусть $s_n = e^{i \omega n}$ и $y_n = H(\omega)e^{i \omega n} $, тогда<br>
$ \begin{cases}
y_{n+1} = H(\omega)e^{i \omega n} + \frac{e^{i \omega n}+ e^{i \omega (n+1)}}{2} \\
y_{n+1} = H(\omega)e^{i \omega (n+1)}
\end{cases}$ <br>

$H(\omega)(e^{i \omega n}e^{i \omega}) = H(\omega)e^{i \omega n} + e^{i \omega n}\frac{1 + e^{i \omega}}{2}$ <br>

$H(\omega)(e^{i \omega} - 1) = \frac{1+e^{i \omega}}{2}$ <br>

$H(\omega) = \frac{1+e^{i \omega}}{2(e^{i \omega} -1)} = \frac{\cos{\frac{\omega}{2}}}{2i\sin{\frac{\omega}{2}}}$ <br>

$\widetilde{H}(f) = \frac{\cos{(\pi f)}}{2i \sin{(\pi f)}}$ <br>

Точное значение интеграла $e^{i \omega t}$ равно $\frac{e^{i \omega t}}{i \omega}$:<br>
$\gamma = \frac{Вычисленное}{Точное} = \cos{\frac{\omega}{2}} \frac{\frac{\omega}{2}}{\sin{(\frac{\omega}{2})}} = 1 - \frac{\omega^2}{12}+ \frac{\omega^4}{720} + ...$ <br> 

Формула Симпсона: <br>
$y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1}+4s_n+s_{n+1}), y_0=0 $ <br>

Пусть $s_n= e^{i \omega n}$ и $y_n = H(\omega)e^{i \omega n} $, тогда <br>
$ \begin{cases}
y_{n+1} = H(\omega)e^{i \omega n} + \frac{e^{i \omega (n-1)}+ 4e^{i \omega n}+ e^{i \omega (n+1)}}{3} \\
y_{n+1} = H(\omega)e^{i \omega (n+1)}
\end{cases}$ <br>

$H(\omega)(e^{i \omega n}e^{i \omega}) = H(\omega)e^{i \omega n}e^{-i\omega} + e^{i \omega n}\frac{e^{-i \omega}+4+e^{i\omega}}{3}$ <br>

$H(\omega)(e^{i \omega} - e^{-i \omega}) = \frac{e^{-i \omega}+4+e^{i \omega}}{3}$ <br>

$H(\omega) = \frac{e^{-i \omega}+4+e^{i \omega}}{3(e^{i \omega} - e^{-i \omega})} = \frac{\cos{\omega} +2}{3i\sin{\omega}} $ <br>

$\widetilde{H}(f) = \frac{\cos{(2\pi f)} +2}{3i \sin{(2\pi f)}}$ <br>

Точное значение интеграла $e^{i \omega t}$ равно $\frac{e^{i \omega t}}{i \omega}$:<br>
$\gamma = \frac{Вычисленное}{Точное} = \frac{(\cos{\omega} +2)i \omega}{3i\sin{\omega}} = \frac{\cos{\omega} +2}{3} * \frac{\omega}{\sin{\omega}} = 1+ \frac{\omega^4}{180} + ...$ <br> 


```python
import numpy as np
import math
import matplotlib.pyplot as plt

h_rect = lambda x: (1 / (2j*np.sin(math.pi*x))).imag
h_trap = lambda x: (np.cos(math.pi*x) / (2j*np.sin(math.pi*x))).imag
h_simp = lambda x: ((np.cos(2*math.pi*x)+2)/
(3j*np.sin(2*math.pi*x))).imag
y_rect = lambda x: math.pi*x / (np.sin(math.pi * x))
y_trap = lambda x: np.cos(math.pi * x)*(math.pi*x/np.sin(x*math.pi))
y_simp = lambda x: (np.cos(2*math.pi*x) + 2)/3
t = np.linspace(1e-10, 0.5, 300)
plt.plot(t, h_rect(t))
plt.xlabel('t')
plt.ylabel(r'$H(t)$')
plt.title('График отношения вычисляемого в результате фильтрации значения к истинному для формул прямоугольников.' )
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -50, 10.))
plt.show()
plt.plot(t, h_trap(t))
plt.xlabel('t')
plt.ylabel(r'$H(t)$')
plt.title('График отношения вычисляемого в результате фильтрации значения к истинному для формул трапеций' )
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -50, 10.))
plt.show()
plt.plot(t, h_simp(t))
plt.xlabel('t')
plt.ylabel(r'$H(t)$')
plt.title('График отношения вычисляемого в результате фильтрации значения к истинному для формул Симпсона' )
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -50, 10.))
plt.show()
plt.plot(t, y_rect(t), label=r'$y_{rect}$')
plt.plot(t, y_trap(t), label=r'$y_{trap}$')
plt.plot(t, y_simp(t), label=r'$y_{simp}$')
plt.xlabel('t')
plt.ylabel(r'$y(t)$')
plt.legend()
plt.title('Графики отношения вычисляемого в результате фильтрации значения к истинному для формул прямоугольников, для формул трапеций и для формул Симпсона' )

plt.show()
```

На построенных графиках передаточных функций видно, что все три фильтра одинаково снижают уровень низких частот. Однако их реакция на высокие частоты различается: фильтр на основе формулы трапеций полностью гасит высокие частоты; фильтр с формулой прямоугольника демонстрирует слабое отрицательное усиление; а у фильтра Симпсона график симметричен, и функция стремится к минус бесконечности как на низких, так и на высоких частотах.

### Задание 2
Вывести формулу передаточной функции рекурсивного фильтра для интегрирования «по правилу 3/8»: <br>
$ y_{n +2} = y_{n-1} + \frac{1}{8}(x_{n+2} +3x_{n+1} + 3x_n + x_{n-1}) \\$
Построить график передаточной функции и график отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства передаточной функции.



Пусть $ x_{n} = e^{iwn} $, $ y_{n} = H(w)*e^{iwn}  \\$
$
\begin{equation}
 \begin{cases}
   y_{n+2} = H(w)e^{iw(n - 1)} + \frac{1}{8}(e^{iw(n+2)} + 3e^{iw(n+1)} + 3e^{iwn} + e^{iw(n-1)}) 
   \\
   y_{n+2} = H(w)e^{iw(n+2)}
 \end{cases}
\end{equation}
\\
$

$
H(w)e^{iw(n+2)} = H(w)e^{iw(n-1)} + \frac{1}{8}(e^{iw(n+2)} + 3e^{iw(n+1)} + 3e^{iwn} + e^{iw(n-1)})
\\
H(w)(e^{2iw} - e^{-iw}) = \frac{1}{8}(e^{2iw} + 3e^{iw} + 3 + e^{-iw})
\\
H(w) = \frac{e^{2iw} + 3e^{iw} + 3 + e^{-iw}}{8(e^{2iw} - e^{-iw})} = \frac{e^{\frac{3iw}{2}} + 3e^{\frac{iw}{2}} + 3e^{\frac{-iw}{2}} + e^{\frac{-3iw}{2}}}{8(e^{\frac{3iw}{2}} - e^{\frac{-3iw}{2}})} = \frac{2cos(\frac{3w}{2}) + 6cos(\frac{w}{2})}{16isin(\frac{3w}{2})} = \frac{cos(\frac{3w}{2}) + 3cos(\frac{w}{2})}{8isin(\frac{3w}{2})}
\\
$

Точное значение интеграла
$
\lambda = \frac{Вычисляемое}{Точное} = \frac{(cos(\frac{3w}{2}) + 3cos(\frac{w}{2}))iw}{8isin(\frac{3w}{2})} = \frac{(cos(\frac{3w}{2}) + 3cos(\frac{w}{2}))w}{8sin(\frac{3w}{2})} = \frac{1}{12}(cos(\frac{3w}{2}) + 3cos(\frac{w}{2}))*\frac{\frac{3w}{2}}{sin(\frac{3w}{2})}
\\
$


```python
H = lambda x: ((np.cos(3*2*math.pi*x / 2) + 3*np.cos(2*math.pi*x/2)) / (8j* np.sin(3*2*math.pi*x / 2))).imag
a = lambda x: 1/12 * (np.cos(3 * 2*math.pi*x / 2) + 3 * np.cos(2*math.pi*x / 2)) * ((3* 2*math.pi*x / 2) / (np.sin(3*2*math.pi*x/2)))
t = np.linspace(1e-2, 0.5, 300)
plt.plot(t, H(t))
plt.xlabel('t')
plt.ylabel(r'$H(t)$')
plt.title('График передаточной функции рекурсивного фильтра для интегрирования по правилу "3/8"' )
plt.show()


plt.plot(t, a(t))
plt.xlabel('t')
plt.ylabel(r'$ɣ$')
plt.title('График отношения вычисляемого в результате фильтрации значения к истинному' )
plt.show()

```
