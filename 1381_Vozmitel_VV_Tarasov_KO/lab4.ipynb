1. Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона. Построить графики передаточных функций и графики отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства полученных передаточных функций.

Формула прямоугольника: $y_{n+1} = y_{n} + s_{n+\frac{1}{2}}, y_{0} = 0$

&emsp;&emsp;Пусть $s_{n} = e^{i\omega n}$ и $y_{n} = H(\omega)e^{i\omega n}$, тогда: $\begin{cases} y_{n+1} = H(\omega)e^{i\omega n} + e^{i\omega n}e^{\frac{1}{2} i\omega} \\ y_{n+1} = H(\omega)e^{i\omega (n+1)} \end{cases}$


&emsp;&emsp;$ H(\omega)(e^{i\omega n}e^{i\omega} - e^{i\omega n}) = e^{i\omega n}e^{\frac{1}{2} i\omega} $

&emsp;&emsp;$H(\omega)(e^{i\omega} - 1) = e^{\frac{1}{2} i\omega}$

&emsp;&emsp;$H(\omega) = \frac{1}{e^{\frac{i\omega}{2}} - e^{\frac{-i\omega}{2}}} = \frac{1}{2isin(\frac{\omega}{2})}$

&emsp;&emsp;$\widetilde{H}(f) = \frac{1}{2isin(\pi f)}$

&emsp;&emsp;Точное значение интграла от $e^{i\omega n}$ равно $\frac{e^{i\omega n}}{i\omega}$

&emsp;&emsp;Отношение значений: $\frac{Вычисленное}{Точное} = \frac{\frac{\omega}{2}}{sin(\frac{\omega}{2})} = 1 + \frac{x^2}{24} + \frac{7x^4}{5760} + ... = \frac{\pi f}{sin(\pi f)} = 1 + \frac{\pi^2f^2}{6} + \frac{7 \pi^4f^4}{360} + ...$


Формула трапеций: $y_{n+1} = y_{n} + \frac{1}{2}(s_{n} + s_{n+1}), y_{0} = 0$

&emsp;&emsp;Тогда по аналогии с формулой прямоугольника получаем: $\begin{cases} y_{n+1} = H(\omega)e^{i\omega n} + \frac{e^{i\omega n} + e^{i\omega (n+1)}}{2} \\ y_{n+1} = H(\omega)e^{i\omega (n+1)} \end{cases}$

&emsp;&emsp;$H(\omega)(e^{i\omega n}e^{i\omega}) = H(\omega)e^{i\omega n} + e^{i\omega n} \frac{1 + e^{i\omega}}{2}$

&emsp;&emsp;$H(\omega)(e^{i\omega} - 1) = \frac{1 + e^{i\omega}}{2}$

&emsp;&emsp;$H(\omega) = \frac{1+e^{i\omega}}{2(e^{i\omega} - 1)} = \frac{cos(\frac{\omega}{2})}{2isin(\frac{\omega}{2})}$

&emsp;&emsp;$\widetilde{H}(f) = \frac{cos(\pi f)}{2isin(\pi f)}$

&emsp;&emsp;Точное значение интграла от $e^{i\omega n}$ равно $\frac{e^{i\omega n}}{i\omega}$

&emsp;&emsp;Отношение значений: $\frac{Вычисленное}{Точное} = cos(\frac{\omega}{2}) \frac{\frac{\omega}{2}}{sin(\frac{\omega}{2})} = 1 - \frac{\omega^{2}}{12} + \frac{\omega^{4}}{720} + ... = cos(\pi f)\frac{\pi f}{sin(\pi f)} = 1 - \frac{\pi^{2} \omega^{2}}{3} + \frac{\pi^{4} \omega^{4}}{45} + ...$

Формула Симпсона: $y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_{n} + s_{n+1}), y_{0} = 0$

&emsp;&emsp;По аналогии с предыдущими пунктами имеем: $\begin{cases} y_{n+1} = H(\omega)e^{i\omega (n-1)} + \frac{e^{i\omega (n-1)} + 4e^{i\omega n)} + e^{i\omega (n+1)}}{3} \\ y_{n+1} = H(\omega)e^{i\omega (n+1)} \end{cases}$

&emsp;&emsp;$H(\omega)(e^{i\omega n}e^{i\omega}) = H(\omega)e^{i\omega n}e^{-i\omega} + e^{i\omega n} \frac{e^{-i\omega} + 4 + e^{i\omega}}{3}$

&emsp;&emsp;$H(\omega) = \frac{e^{-i\omega} + 4 + e^{i\omega}}{3(e^{i\omega} - e^{-i\omega})} = \frac{cos(\omega)+2}{3isin(\omega)}$

&emsp;&emsp;$\widetilde{H}(f) = \frac{cos(2\pi f) + 2}{3isin(2\pi f)}$

&emsp;&emsp;Точное значение интграла от $e^{i\omega n}$ равно $\frac{e^{i\omega n}}{i\omega}$

&emsp;&emsp;Отношение значений: $\frac{Вычисленное}{Точное} = \frac{(cos(\omega)+2)i\omega}{3isin(\omega)} = \frac{cos{\omega} + 2}{3} * \frac{\omega}{sin{\omega}} = 1 + \frac{\omega^{4}}{180} + ... = \frac{cos(2\pi f) + 2}{3} * \frac{2\pi f}{sin(2\pi f)} = 1 + \frac{4\pi^{4}f^{4}}{45} + ...$

```python
# График передаточной функции для формулы прямоугольника
import numpy as np
import matplotlib.pyplot as plt

def rect(f):
    return (1 / (2j*np.sin(np.pi*f))).imag

f =  np.linspace(1e-10, 0.5, 300, endpoint=True)

plt.plot(f, rect(f), color="red", linewidth=2)
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -30, 10))
plt.title("График передаточной функции для формулы прямоугольника")
plt.xlabel('f')
plt.ylabel('H(f)')
```

```python
# График отношения вычисляемого в результате фильтрации значения к истинному для формулы прямоугольника
import numpy as np
import matplotlib.pyplot as plt

np.seterr(divide='ignore', invalid='ignore')
def rect(f):
    return (np.pi * f) / (np.sin(np.pi*f))

f =  np.linspace(0, 0.5, 300, endpoint=True)

plt.plot(f, rect(f), color="red", linewidth=2)

plt.title("График отношения для формулы прямоугольника")
plt.xlabel('f')
plt.ylabel('y')
```

```python
# График передаточной функции для формулы трапеции
import numpy as np
import matplotlib.pyplot as plt

def trap(f):
    return ( np.cos(np.pi * f) / (2j*np.sin(np.pi*f)) ).imag

f =  np.linspace(1e-10, 0.5, 300, endpoint=True)

plt.plot(f, trap(f), color="red", linewidth=2)
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -30, 10))
plt.title("График передаточной функции для формулы трапеции")
plt.xlabel('f')
plt.ylabel('H(f)')
```

```python
# График отношения вычисляемого в результате фильтрации значения к истинному для формулы трапеции
import numpy as np
import matplotlib.pyplot as plt

np.seterr(divide='ignore', invalid='ignore')
def rect(f):
    return np.cos(np.pi * f) * ( (np.pi * f) / (np.sin(np.pi*f)) )

f =  np.linspace(0, 0.5, 300, endpoint=True)

plt.plot(f, rect(f), color="red", linewidth=2)

plt.title("График отношения для формулы трапеции")
plt.xlabel('f')
plt.ylabel('y')
```

```python
# График передаточной функции для формулы Симпсона
import numpy as np
import matplotlib.pyplot as plt

def trap(f):
    return ( (np.cos(2*np.pi * f) + 2) / (3j*np.sin(2*np.pi*f)) ).imag

f =  np.linspace(1e-10, 0.5, 300, endpoint=True)

plt.plot(f, trap(f), color="red", linewidth=2)
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -30, 10))
plt.title("График передаточной функции для формулы Симпсона")
plt.xlabel('f')
plt.ylabel('H(f)')
```

```python
# График отношения вычисляемого в результате фильтрации значения к истинному для формулы Симпсона
import numpy as np
import matplotlib.pyplot as plt

np.seterr(divide='ignore', invalid='ignore')
def rect(f):
    return  (np.cos(2*np.pi * f) + 2) / 3 * ((2*np.pi * f) / np.sin(2*np.pi*f))

f =  np.linspace(0, 0.5, 300, endpoint=True)

plt.plot(f, rect(f), color="red", linewidth=2)

plt.title("График отношения для формулы Симпсона")
plt.xlabel('f')
plt.ylabel('y')
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, 0, 20))
```

2. Вывести формулу передаточной функции рекурсивного фильтра для интегрирования «по правилу 3/8»: $y_{n+2} = y_{n-1} + \frac{1}{8}(x_{n+2} + 3x_{n+1} + 3x_{n} + x_{n-1}) $

    Построить график передаточной функции и график отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства передаточной функции.

    

Пусть $s_{n} = e^{i\omega n}$ => $y_{n} = H(\omega)e^{i\omega n}$ :

$\begin{cases} y_{n+2} = H(\omega)e^{i\omega (n-1)} + \frac{e^{i\omega (n+2)} + 3e^{i\omega (n + 1)} + 3e^{i\omega n} + e^{i\omega (n - 1)}}{8} \\ y_{n+2} = H(\omega)e^{i\omega (n+2)} \end{cases}$

$H(\omega)(e^{i\omega n}e^{2i\omega}) = H(\omega)e^{i\omega n}e^{-i\omega} + e^{i\omega n} \frac{e^{2i\omega } + 3e^{i\omega} + 3 + e^{-i\omega}}{8}$

$H(\omega)(e^{2i\omega } - e^{-i\omega}) = \frac{e^{2i\omega } + 3e^{i\omega} + 3 + e^{-i\omega}}{8}$

$H(\omega) = \frac{e^{2i\omega } + 3e^{i\omega} + 3 + e^{-i\omega}}{8(e^{2i\omega } - e^{-i\omega})} * \frac{e^{\frac{-i\omega}{2}}}{e^{\frac{-i\omega}{2}}}$

$H(\omega) = \frac{{e} ^ {\frac{3i\omega}{2}} +3 {e} ^ {\frac{i\omega}{2}} + {3e} ^ {\frac{-i\omega}{2}} + {e} ^ {\frac{-3i\omega}{2}}}{8( {e} ^ {\frac{3i\omega}{2}} - {e} ^ { \frac{-3i\omega}{2}} )} = \frac{2cos \frac{3\omega}{2} +6cos \frac{\omega}{2}}{16isin \frac{3\omega}{2}} $

$\tilde{H}(f) = \frac{cos {(3πf)} +3 cos⁡ (πf)}{8isin(3πf)}$

Значение интеграла $e^{i\omega t} = \frac{e^{i\omega t}}{i \omega}$ тогда:

$γ= \frac {Вычисленное}{Точное} = \omega \frac{cos {\frac{3 \omega}{2}} +3cos \frac {\omega}{2}}{8sin \frac{3 \omega}{2}} = \frac{1}{12}(cos {\frac{3 \omega}{2}} +3cos \frac{\omega}{2}) * \frac{\frac{3 \omega}{2}}{sin \frac{3 \omega}{2}} = \frac{1}{3} + \frac{{\omega} ^ {4}}{240} + … $

$γ= \frac{1}{12} ( cos {(3πf)} +3 cos {(πf)} ) \frac{3πf}{sin(3πf)} = \frac{1}{3} + \frac{{π} ^ {4} {f} ^ {4}}{15} + …$



```python
# График передаточной функции
import numpy as np
import matplotlib.pyplot as plt

def rect(f):
    return ((np.cos(3 * np.pi * f) + 3 * np.cos(np.pi * f)) / (8j*np.sin(3 * np.pi*f))).imag

f =  np.linspace(1e-10, 0.5, 300, endpoint=True)

plt.plot(f, rect(f), color="red", linewidth=2)
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -30, 10))
plt.title("График передаточной функции")
plt.xlabel('f')
plt.ylabel('H(f)')
```

```python
# График отношения вычисляемого в результате фильтрации значения к истинному
import numpy as np
import matplotlib.pyplot as plt

np.seterr(divide='ignore', invalid='ignore')
def rect(f):
    return 1/12 * (np.cos(3 * np.pi * f) +3 * np.cos(np.pi * f)) * ((3 * np.pi * f)/(np.sin(3 * np.pi * f)))

f =  np.linspace(0, 0.5, 300, endpoint=True)

plt.plot(f, rect(f), color="red", linewidth=2)

plt.title("График отношения для формулы прямоугольника")
plt.xlabel('f')
plt.ylabel('y')
```
