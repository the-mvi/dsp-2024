3. Смоделировать дискретную экспоненциальную функцию $s_{1}(k)$ с выводом графиков на интервале дискретного времени $nT\in \left[ 0, (N-1)T \right]$ и дискретного нормированного времени $n \in \left[ 0, N - 1 \right]$.

    Пояснить соответствие между дискретной и аналоговой экспонентами.

    $s_{1}(k)$ = $\begin{cases} 0, n < 0; \\ -0.815^{n}, n\ge 0 \end{cases}$

```python
#Дискретная экспоненциальная функция на интервале дискретного времени
import numpy as np
import matplotlib.pyplot as plt

N = 33
T = 0.0005
a = -0.815
C = 4
w0 = np.pi / 9
m = 8

x = np.linspace(0, (N-1)*T, N, endpoint=True)
def exp(n):
    return np.power(a, n)


for i in range(N):
    y = exp(i)

    plt.scatter(x[i], y, color='red')
    plt.plot([x[i], x[i]], [0, y], color="green")
```

```python
#Дискретная экспоненциальная функция на интервале дискретного нормированного времени
import numpy as np
import matplotlib.pyplot as plt

N = 33
T = 0.0005
a = -0.815
C = 4
w0 = np.pi / 9
m = 8

x = np.linspace(0, (N-1), N, endpoint=True)
def exp(n):
    return np.power(a, n)


for i in range(N):
    y = exp(i)

    plt.scatter(x[i], y, color='red')
    plt.plot([x[i], x[i]], [0, y], color="green")
```

    &emsp;$\delta_{d}(n - 8)$ = $\begin{cases} 1, n = 8; \\ 0, n\neq 8 \end{cases}$

    &emsp;$\sigma_{d}(n - 8)$ = $\begin{cases} 1, n \ge  8; \\ 0, n \lt 8 \end{cases}$

```python
#Дискретная экспоненциальная функция со сдвигом 8
import numpy as np
import matplotlib.pyplot as plt

N = 33
T = 0.0005
a = -0.815
C = 4
w0 = np.pi / 9
m = 8

x = np.linspace(0, (N-1)*T, N, endpoint=True)
def exp(n):
    if (n < 8):
        return 0
    else:
        return np.power(a, n - m)


for i in range(N):
    y = exp(i)

    plt.scatter(x[i], y, color='red')
    plt.plot([x[i], x[i]], [0, y], color="green")
```
