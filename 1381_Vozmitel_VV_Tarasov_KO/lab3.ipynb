## Основные теоретические положения

Дискретный нерекурсивный фильтр описывает выходной сигнал y(n) как последовательность чисел на основе входного детерминированного сигнала x(n). Одним из способов представления этой формулы может быть следующее:

$$y_n = \sum_{k=0}^{m} c_k x_{n-k}$$

Передаточная функция - это мат выражение, которое показывает, как система преобразует входной сигнал в выходной. Для нерекурсивного фильтра ее можно найти следующим образом:

$$H(w) = \frac{y_n}{x_n}$$

Если сигнал гармонический дискретный: $$ H(w) = \frac{ \sum_{k=0}^{m} c_k e^{i\omega(n-k)} }{ e^{iwn} } = \sum_{k=0}^{m} c_k e^{-i\omega k} $$

Процедура, известная как полиномиальное сглаживание, представляет собой метод аппроксимации значений конечного числа элементов ряда с использованием полинома определенной степени по методу наименьших квадратов. Значение сглаживающего полинома присваивается среднему значению элементов в центре выбранного временного интервала.

Формулы Спенсера:

    Входной сигнал: $s(t)$

    Выходной сигнал: $y(t) = A + Bt$

    Приближение по прямой линии по $k$ точкам: $F(A, B) = \sum_{m = -k}^{k}(s_{m} - y_{m})^2 = \sum_{m = -k}^{k}(s_{m} - A - Bm)^2 \Rightarrow min$

    Чтобы найти минимум, нужно продифференцировать по $A$ и $B$ и приравнять к нулю:

    
    &emsp;&emsp;$ \begin{cases} \frac{\delta F(A, B)}{\delta A} = 0 \\ \frac{\delta F(A, B)}{\delta B} = 0 \end{cases} \Leftrightarrow \begin{cases} -2 \sum_{m=-k}^{k}(s_{m} - A - Bm) = 0 \\ -2\sum_{m=-k}^{k}(ms_{m} - mA - Bm^2) = 0  \end{cases} \Leftrightarrow \begin{cases} - \sum_{m = -k}^{k} s_{m} + \sum_{m = -k}^{k} A + \sum_{m = -k}^{k} Bm = 0 \\ - \sum_{m = -k}^{k} ms_{m} + \sum_{m = -k}^{k} mA + \sum_{m = -k}^{k} Bm^2 = 0 \end{cases}$

    Тогда система нормальных уравнений получается следующей:

    &emsp;&emsp;$\begin{cases} A = \frac{1}{2k + 1} \sum_{m=-k}^{k} s_{m} \\ B = \frac{\sum_{m=-k}^{k}ms_{m}}{ \sum_{m=-k}^{k}m^2}\end{cases}$

    Получаем: $y_{0} = A = \frac{1}{2k + 1} \sum_{m=-k}^{k} s_{m} = \frac{1}{2k + 1} \sum_{m=-k}^{k}(s_{-k} + s_{-k+1} + ... + s_{k-1} + s_{k})$

    В общем случае ($s_{n} = e^{i \omega n}$):
    
    &emsp;&emsp;$y_{n} = \frac{1}{2k + 1} \sum_{m=-k}^{k}(s_{-k+n} + s_{-k+1+n} + ... + s_{k-1+n} + s_{k+n}) = \frac{1}{2k + 1} \sum_{m=-k}^{k}(e^{-k i \omega} + e^{(-k + 1) i \omega} + ... + e^{(k - 1)i \omega} + e^{ k i \omega }) = H(\omega)$

    Используя следствие формулы Эйлера - $ e^{ix} + e^{-ix} = 2 \cos(x) $, получим:

    &emsp;&emsp;$H(\omega) = \frac{1}{2k + 1} \sum_{m=-k}^{k}(1 + 2cos(\omega) + 2cos(2\omega) + 2cos(k\omega))  = \frac{sin(\frac{(2k+1)\omega)}{2}))}{(2k+1)sin(\frac{\omega}{2})))} = H(2\pi f) = \widetilde{H}(f)$

    Таким образом получаются следующие формулы:
    
    &emsp;&emsp;$\widetilde{H}_{3}(f) = \frac{sin(3\pi f)}{3sin(\pi f)}$; $\widetilde{H}_{5}(f) = \frac{sin(5\pi f)}{5sin(\pi f)}$, $\widetilde{H}_{7}(f) = \frac{sin(7\pi f)}{7sin(\pi f)}$, $\widetilde{H}_{9}(f) = \frac{sin(9\pi f)}{9sin(\pi f)}$

    
   

Построим графики функций на интерваоле $f \in \left( -0.5, 0.5 \right)$
